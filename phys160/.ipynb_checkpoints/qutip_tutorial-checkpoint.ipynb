And the state $|1\rangle$

```python
one = basis(2,1)
one
```

If we are really lazy and don't want to type out the $\sqrt{2}$ for normalization, we can use the fact that quantum objects have a number of handy methods associated with them. One of them, Qobj.unit(), returns a normalized version of the quantum object in question

```python
plus_unnorm = zero + one
plus = plus_unnorm.unit()
plus
```

And of course, if we need to describe states using density matrices, this can be done. In fact, we can easily write down a pure state $$\rho = |\psi \rangle \langle \psi|$$ with the ket2dm command

```python
plus_dm = ket2dm(plus) #Takes in a state (ket) and makes it a density matrix
plus_dm
```

It's also easy to write down operators that we know and love, such as the Pauli matrices. For a list of operators you can write down easily, see http://qutip.org/docs/latest/apidoc/functions.html#module-qutip.operators

## Visualizing quantum objects

One of the nice aspects about QuTiP is how easy it is to visualize quantum objects on the Bloch sphere. This can be done by instantiating a Bloch object and adding points and vectors to it. For a reference, see http://qutip.org/docs/4.1/guide/guide-bloch.html and http://qutip.org/docs/4.0.2/guide/guide-states.html

```python
# Directly add a quantum object to the Bloch sphere
bloch_sphere = Bloch() #Starts the Bloch sphere
bloch_sphere.add_states([plus,one]) #Adds a state or set of states.
bloch_sphere.show()
```

We can also add points to the Bloch sphere by simply providing [x, y, z] coordinates. And by providing arrays of coordinates, we can plot trajectories. Shown below is a dummy trajectory going from +x to -x through the middle of the sphere.

```python
# Add a trajectory to the Bloch sphere
bloch_sphere = Bloch()
x = np.linspace(-1, 1, 100)
y = np.zeros(100)
z = np.zeros(100)
bloch_sphere.add_points([x, y, z], meth='l') #Adds points. Note that x[i],y[i],z[i] are the ith point.
bloch_sphere.show()
```

### Quantum unitary evolution

In order to simulate unitary quantum dynamics, we typically have to follow the steps:
1. Write down a Hamiltonian that describes how the system coherently evolves. For example: $$H = \Delta \sigma_z + \Omega \sigma_x$$
2. Write down differential equations for each amplitude $c_{i}$ using the Schrodinger equation.
3. Solve the system of coupled differential equations, usually numerically.

Often times, step 2 can be quite time-consuming and tedious. One of the *main advantages* of QuTiP is that we can skip step 2 and obtain a numerical simulation of our open quantum system directly after step 1. Below, we examine how exactly this works.

### Excercise: 
suppose we have the Hamiltonian in 1. above and we start in $|0\rangle$. Compute the dynamics for $\{\Delta,\Omega\}=\{0,1\}$ and $\{\Delta,\Omega\}=\{10,1\}$. Plot the $\langle \sigma_z\rangle$ for each and use this to explain that in the off-resonance case, the state $|1\rangle$ is not populated almost at all.

```python
def solve_se(delta, omega, t=10): 
    """ Solves a single-qubit Schrodinger equation
    
    See HW3 for details on the Hamiltonian
    :param delta: Detuning (units of 1/time)
    :param omega: Rabi frequency (units of 1/time)
    :param gamma: Spontaneous emission rate (units of 1/time)
    :param time: (opt) time to run until
    """
    
    # Define the Hamiltonian
    H = delta*sigmaz() + omega*sigmax()
    m_ops = [sigmax(), sigmay(), sigmaz()]
    
    # Initial state
    psi_0 = basis(2,0)
    
    # Time steps to measure on
    t = np.linspace(0, t, 1000)
    
    # Solve master equation
    solution = sesolve(H, psi_0, t, e_ops=m_ops) #Measures sigmaz() at every time-slice
    
    return solution.expect   
```

```python
import matplotlib.pyplot as plt
#On resonance goes here: 
delta, omega, t = 0, 1, 10

result_res = solve_se(delta, omega, t)

#Off resonance goes here:
delta, omega, t = 10, 1, 100

result_offres = solve_se(delta, omega, t)


plt.figure()
plt.plot(np.linspace(0,t,1000), result_res[0], label='Resonance')
plt.plot(np.linspace(0,t,1000), result_offres[0], label='Off-Resonance')
plt.legend()
plt.show()
```

## Simulating quantum dynamics (non-unitary)

In order to simulate open quantum dynamics, we typically have to follow the steps:
1. Write down a Hamiltonian that describes how the system coherently evolves. For example: $$H = \Delta \sigma_z + \Omega \sigma_x$$
2. Specify how the system interacts with the environment, often in the form of collapse operators $L$. For example: $$L = \sqrt{\gamma} \sigma_z$$
3. Write down differential equations for each density matrix component $\rho_{ij}$ using the master equation.
4. Solve the system of coupled differential equations, usually numerically.

Often times, step 3 can be quite time-consuming and tedious. One of the *main advantages* of QuTiP is that we can skip step 3 and obtain a numerical simulation of our open quantum system directly after steps 1 and 2. Below, we examine how exactly this works.

In section, we solved the problem of a single qubit experiencing decoherence $L=\sqrt{\gamma}\sigma_z$ and in the presence of a Hamiltonian $H=\Delta \sigma_z+\Omega\sigma_x$. Here, we plot our findings starting with the pure state $|0\rangle$. We plot the trajectory of the state in the Bloch sphere using (1) the master equation, and (2) the the equations of motion of the Bloch vector $\vec{r} = (r_x,r_y, r_z)$. 

```python
def solve_me(delta, omega, gamma, time=10):
    """ Solves a single-qubit master equation
    
    See HW3 for details on the Hamiltonian
    :param delta: Detuning (units of 1/time)
    :param omega: Rabi frequency (units of 1/time)
    :param time: (opt) time to run until
    """
    
    # Define the Hamiltonian
    H = delta*sigmaz() + omega*sigmax()
    
    # Collapse (Lindblad) operators
    #
    # NOTE: we use sigmap() here because this performs
    # the desired |0><1| operation mapping state |1>
    # onto state |0>. Check this for yourself!
    #
    
    c_ops = [np.sqrt(gamma)*sigmaz()]
    
    # Operators to measure
    m_ops = [sigmax(), sigmay(), sigmaz()]
    
    # Initial state
    psi_0 = basis(2,0)
    
    # Time steps to measure on
    time = np.linspace(0, time, 1000)
    
    # Solve master equation
    solution = mesolve(H, psi_0, time, c_ops=c_ops, e_ops=m_ops)
    
    return solution.expect
```

Now, all we have to do is specify the Hamiltonian parameters and we can let QuTiP do all the work!

```python
# Parameters
delta = 0
omega = 1 * 2*np.pi
gamma = 0.05 * 2*np.pi

# Time to plot for
time = 10

#Solving it with the master equation
result = solve_me(delta, omega, gamma, time=time)
time_axis = np.linspace(0, time, 1000)

#Solving it with my equations of motion for the bloch vector r which we derived in lecture: 
def bloch_vect_decoh(r, t, delta, omega, gamma):
    rx, ry, rz = r
    dydt = [2*(-gamma*rx+delta*ry), 2*(-gamma*ry-delta*rx+omega*rz), -2*omega*ry]
    return dydt

from scipy.integrate import odeint
sol = odeint(bloch_vect_decoh, [0,0,1], time_axis, args=(delta, omega, gamma))

#Plotting it out
bloch_sphere = Bloch()
bloch_sphere.add_points([result[0],result[1],result[2]], meth='l')
#bloch_sphere.add_points([sol[:, 0],sol[:, 1],sol[:, 2]])
bloch_sphere.show()
```

## Time dependent Hamiltonians

In the previous cases, our Hamiltonians were constant in time. In practical QIP systems, we will apply time-varying control pulses to manipulate our qubit. This can also be treated quite easily in QuTiP by simply specifying the time dependence in a separate function $H_1(t)$. In the case of simple time dependences, we can write our Hamiltonian $$H(t) = H_0 + c(t) H_1$$ where all of the time dependence is specified in $c(t)$. Below, we walk through the case of a linearly increasing Rabi frequency.

See http://qutip.org/docs/latest/guide/dynamics/dynamics-time.html for more details about encoding time-dependent Hamiltonians.

```python
def solve_me_t(delta, omega_function, gamma, time=10):
    """ Solves a single-qubit master equation
    
    See HW3 for details on the Hamiltonian
    :param delta: Detuning (units of 1/time)
    :param omega: Function defining Rabi frequency (units of 1/time)
    :param gamma: Spontaneous emission rate (units of 1/time)
    :param time: (opt) time to run until
    """
    
    # Define the constant Hamiltonian
    H = delta*sigmaz()
    
    # Define time-dependent component
    H1 = sigmax()
    
    # Pass total Hamiltonian as a list
    H = [H, [H1, omega_function]]
    
    # Collapse (Lindblad) operators
    #
    # NOTE: we use sigmap() here because this performs
    # the desired |0><1| operation mapping state |1>
    # onto state |0>. Check this for yourself!
    #
    c_ops = [np.sqrt(gamma)*sigmap()]
    
    # Operators to measure
    m_ops = [sigmax(), sigmay(), sigmaz()]
    
    # Initial state
    psi_0 = basis(2,0)
    
    # Time steps to measure on
    time = np.linspace(0, time, 1000)
    
    # Solve master equation
    solution = mesolve(H, psi_0, time, c_ops=c_ops, e_ops=m_ops)
    
    return solution.expect
```

```python
# Encode time dependence and plot
def omega_coeff(t, args):
    """ Linear ramp
    
    :param t: time
    :param args: unused
    """
    
    return t

time = 10
time_axis = np.linspace(0, time, 1000)
plt.xlabel('Time')
plt.ylabel('$\Omega(t)$')
plt.plot(time_axis, omega_coeff(time_axis, 0)/(2*np.pi))
```

```python
# Set other parameters, solve
delta = 0 * 2*np.pi
gamma = 0.05 * 2*np.pi
result = solve_me_t(delta, omega_coeff, gamma, time=time)

# Plot
plt.xlabel('Time')
plt.ylabel('$<\sigma_z>$')
plt.plot(time_axis, result[2])

# Display trajectory on the Bloch sphere
bloch_sphere = Bloch()
bloch_sphere.add_points([result[0],result[1],result[2]], meth='l')
bloch_sphere.show()
```

### Steady state using QuTip

QuTip also allows us to solve for the steady state density matrix given a Hamiltonian H, and collapse/jump operators c_ops. To do this: 

1. Define the Hamiltonian. For example, we can use again $H=\Delta\sigma_z+\Omega\sigma_x$. 
2. Define the collapse operators. For example, we can use spontaneous emission. 
3. Call rho_ss = steadystate(H, c_ops) to get the steady state. One can then lookup individual terms of the steady state. For example, $\rho_{11} =$ rho_ss[1,1]

```python
def solve_ss(delta, omega, gamma): 
    """ Solves a single-qubit master equation
    
    See HW3 for details on the Hamiltonian
    :param delta: Detuning (units of 1/time)
    :param omega: Function defining Rabi frequency (units of 1/time)
    :param gamma: Spontaneous emission rate (units of 1/time)
    :param time: (opt) time to run until
    """

    H = delta*sigmaz()+omega*sigmax()
    c_ops = [np.sqrt(gamma)*sigmap()]
    
    # Solve master equation
    solution = steadystate(H,c_ops)
    
    return solution
```
