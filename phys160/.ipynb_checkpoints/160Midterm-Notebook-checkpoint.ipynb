```python
# Import necessary packages
from qutip import *
import numpy as np
import matplotlib.pyplot as plt

# Define some constants
zero, one = basis(2, 0), basis(2, 1)
```

### Master Equation

```python
def solve_me(J, gamma, times):
    # Hamiltonian H_3
    one_one = ket2dm(basis(2, 1))
    H = J * tensor(one_one, one_one, one_one)
    
    # Operators to measure
    # when blank it calculates full state - use this option
    m_ops = []
    
    # Jump operators
    L_1 = np.sqrt(gamma / 2) * tensor(sigmaz(), identity(2), identity(2))
    L_2 = np.sqrt(gamma / 2) * tensor(identity(2), sigmaz(), identity(2))
    L_3 = np.sqrt(gamma / 2) * tensor(identity(2), identity(2), sigmaz())
    L_ops = [L_1, L_2, L_3]
    
    # Initial state â†’ |+>|+>|1>
    psi_1 = (zero + one).unit()
    psi_2 = (zero + one).unit()
    psi_3 = one
    psi_0 = tensor(psi_1, psi_2, psi_3)
    
    # Solve master equation
    solution = mesolve(H, psi_0, times, L_ops, m_ops)
    
    return solution.states
```

### Run the Simulation

<font color="blue" size="4">
    As a sanity check, let us run the simulation for $\gamma = 0,$ so there is no dephasing and we should get the maximum fidelity at time $\tau = \pi / J.$
</font>

```python
# Time steps to measure on
time = 6
times_nodeph = np.linspace(0, time, 1000)

# Pick random J and set gamma to zero
J = 1
gamma = 0

# Run the thing
result_nodeph = solve_me(J, gamma, times_nodeph)
```

### Fidelity Calculation (No Dephasing)

```python
z, o = zero, one

# Ideal state is |psi> = (|00> + |01> + |10> - |11>) |1>
psi_id = (tensor(z, z, o)  + tensor(z, o, o) + tensor(o, z, o) - tensor(o, o, o)).unit()
rho_id = ket2dm(psi_id)

fidelities = []
for state in result_nodeph: 
    fidelities.append(fidelity(state, rho_id))
fidelities = np.array(fidelities)

plt.plot(times_nodeph, fidelities)
plt.xlabel('Time', fontsize=16)
plt.ylabel(r'Fidelity $\mathcal{F}$', fontsize=16)
plt.title(r'Fidelity ($\mathcal{F}$) vs. Time for No Dephasing', fontsize=18)
plt.show()
```

<font color="blue" size="4">
    The maximum is indeed at $t = \pi / J \approx 3$ s!
</font>

<font color="blue" size="4">
    Now we will run the evolution for some time picking some random $J/\gamma$ to get a real picture about the effect of the dephasing.
</font>

```python
# Time steps to measure on
time = 10
times_deph = np.linspace(0, time, 1000)

# Pick some random J, gamma
J = 1
gamma = 0.1

# Run the thing
result_deph = solve_me(J, gamma, times_deph)
```

### Fidelity Calculation (Dephasing)

```python
fidelities = []
for state in result_deph: 
    fidelities.append(fidelity(state, rho_id))
fidelities = np.array(fidelities)

plt.plot(times_deph, fidelities)
plt.xlabel('Time', fontsize=16)
plt.ylabel(r'Fidelity $\mathcal{F}$', fontsize=16)
# plt.ylim((0.25, 1))
plt.title(r'Fidelity ($\mathcal{F}$) vs. Time for $J / \gamma = 10$', fontsize=18)
plt.show()
```

<font color="blue" size="4">
    We see that the dephasing changes how the fidelity advances with time, and the maximum of the fidelity no longer occurs at $\tau = \pi / J.$
</font>

<font color="blue" size="4">
    Now we run the simulation for some time for a series of different values of $J/\gamma.$ In each evolution, we save the maximum fidelity achieved in the chosen time, and then plot this maximum fidelity vs. the values of $J/\gamma$ chosen.
</font>

```python
times = np.linspace(0, 60, 100)
results = []
J_over_gammas = np.linspace(0.1, 10, 100)

for J_over_gamma in J_over_gammas:
    J, gamma = 1, 1 / J_over_gamma
    results.append(solve_me(J, gamma, times))
```

```python
# Fidelity calculation
fidelities = []
for result in results:
    current_fidelities = []
    for state in result:
        current_fidelities.append(fidelity(state, rho_id))
    fidelities.append(np.max(current_fidelities))
fidelities = np.array(fidelities)
```

```python
plt.plot(J_over_gammas, fidelities)
plt.xlabel(r'$J / \gamma$', fontsize=16)
plt.ylabel(r'Fidelity $\mathcal{F}$', fontsize=16)
plt.title(r'Fidelity ($\mathcal{F}$) vs. $J / \gamma$', fontsize=18)
plt.show()
```

<font color="blue" size="4">
    This plot makes sense. When $J / \gamma = 0,$ the dephasing is the dominant contribution, so the best we can hope for is a "random guess," namely, $\mathcal{F} = 1/2.$ And the fidelity should get closer and closer to 1 as $J / \gamma$ increases since then in the limit $\gamma \to 0,$ $J / \gamma \to \infty$ and we recover the original situation with no dephasing and fidelity of 1.
</font>

```python
plt.plot(J_over_gammas, fidelities)
plt.plot(J_over_gammas, np.exp(-2 * np.pi * (1 / J_over_gammas)))
plt.xlabel(r'$J / \gamma$', fontsize=16)
plt.ylabel(r'Fidelity $\mathcal{F}$', fontsize=16)
plt.title(r'Fidelity ($\mathcal{F}$) vs. $J / \gamma$', fontsize=18)
plt.show()
```

<font color="blue" size="4">
    We see very good agreement in the general shape of the graph, with fidelity increasing as $J / \gamma$ increases, which corresponds to $\gamma$ decreasing. However, the actual fidelity starts off at 0.5 whereas our estimate starts at 0. This makes sense because in our estimate we only consider the probability that there is no dephasing error at all as what gives the fidelity, but in reality, there are many other possible scenarios which still recover the desired state, such as an even number of dephasing errors on the same qubit, because then that qubit is not affected (recall $Z^{2n} = \mathbb{1}^n = \mathbb{1}$). So in reality, the fidelity is not just given by the probability of there being no error at allâ€”there are other contributions to it coming from the other possibilities of recovering the ideal state. All of these contributions will simply raise the orange curve above, and the general shape will be similar.
</font>

### Grover's Algorithm

<font color="blue" size="4">
    Let us draw the circuit
</font>

```python
import qiskit as qk
from qiskit.circuit.library import MCMT
from qiskit import Aer
```

```python
# Circuit with three qubits and three classical bits
circ = qk.QuantumCircuit(3, 3)

# Gates
circ.h(0); circ.h(1)
circ += MCMT('cz', 2, 1)
circ.h(0); circ.h(1); circ.h(2)
circ.z(0); circ.z(1)
circ.cz(0, 1)
circ.h(0); circ.h(1)

# Measurement
circ.measure([0, 1], [0,1])

# Draw
circ.draw('mpl')
```
