Here, we will explore two-qubit dynamics in a Markovian noise environment that introduces a dephasing channel on both qubits at the same rate $\Gamma$. Similarly to HW5, we consider the case where the two qubits evolve under a Hamiltonian $J Z_A \otimes Z_B$ in the presence of the dephasing noise, starting in the initial state $\ket{\Psi(0)} = \ket{+_x}_A \otimes \ket{+_x}_B$. 

Our first goal is to explore under which conditions of $J$ relative to $\Gamma$ that the qubits can become entangled. We have already analyzed the effect of this Hamiltonian in problem 3 of Homework 5 without any dephasing. After the qubits evolve from the initial state $\ket{+_x}_A \otimes \ket{+_x}_B$ under the Hamiltonian $H$ for time $Jt = \pi/4$, without dephasing, they became entangled. We will first explore the effect of dephasing noise during the free evolution.

Next, we will quantify the degree of entanglement using an *entanglement witness*. As we saw on HW5, after evolution for time $Jt = \pi/4$, and subsequent single qubit rotations, the qubits are in a Bell state. We will use the fidelity of the Bell state, to quantify the degree of entanglement with and without dephasing. 

## Problem 3a

First, we will define the master equation but set the dephasing rate to zero.

```python
# Import necessary packages
from qutip import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
def solve_me(J, Gamma, times):
    
    """ Solves two qubit dynamics under the master equation
    
    :param J: strength of two-qubit Z_A Z_B interaction
    :param Gamma: rate of dephasing on both qubits
    :param t: times to record the state vector
    
    """
    
    # create Hamiltonian - two qubit interaction H = J * Z_A Z_B
    H = J * tensor(sigmaz(), sigmaz())
    
    # Operators to measure
    # when blank it calculates full state - use this option
    m_ops = []
    
    # Collapse (Lindblad) operators
    c_A = np.sqrt(Gamma) * tensor(sigmaz(), identity(2))# Lindblad operator for qubit A
    c_B = np.sqrt(Gamma) * tensor(identity(2), sigmaz())# Lindblad operator for qubit B
    c_ops = [c_A, c_B]
    
    # Initial state - both qubits along x axis
    psi_A = (basis(2,0)+basis(2,1)).unit()
    psi_B = psi_A
    psi_0 = tensor(psi_A, psi_B)
    
    # Solve master equation
    solution = mesolve(H, psi_0, times, c_ops, m_ops)
    
    return solution.states
```

Let's explore the two-qubit dynamics with the dephasing rate set to zero. Let's choose $J = 2\pi$.

```python
J = 1 * 2*np.pi
```

We extract the final state after solving the master equation, and plot the x component of the Bloch vector of qubit A, as well as the length of the Bloch vector of qubit A. (We can check that the y and z components of the Bloch vector in this case are always zero).

```python
# Solve master equation

# time steps to record the state
t = 10/J # maximum time 
pts = 1000 # number of points
t = np.linspace(0, t, pts)

result = solve_me(J, 0, t) # Gamma is set to zero

# reduced density matrix of qubit A for all time points
rho_A = [result[i].ptrace(0) for i in range(len(result))]

# calculate the bloch vector of qubit A for all time points
bloch_no_dphas = expect(sigmax(), rho_A)

# check that the y and z components of the Bloch vector are still zero
assert np.max(np.abs([expect(sigmay(), rho_A), expect(sigmaz(), rho_A)])) == 0

# plot the x component of the Bloch vector
fig, ax = plt.subplots(1, 2, figsize = (10, 4))
xlabel = 'time'
ax[0].plot(t, bloch_no_dphas, label='$<\sigma_x>_A, \Gamma = 0$')
ax[0].set_xlabel(xlabel)
ax[0].set_ylabel('expectation value of qubit A observable')
ax[0].legend()

# plot the length of the Bloch vector
ax[1].plot(t, np.abs(bloch_no_dphas), label='$\Gamma = 0')
ax[1].set_xlabel(xlabel)
ax[1].set_ylabel('length of Bloch vector')
plt.show()
```

## Problem 3c

Now we will use an entanglement witness to quantify the degree of entanglement between qubits A and B. Applying the unitary $U = e^{i\pi\sigma_x^{(A)}/4}$ on qubit A to the state $\ket{\Psi(t = \frac{\pi}{4J})}$ yields the Bell state $\ket{\Psi^+}$, as you may have seen on Homework 5.

In this section, we will calculate the fidelity of our state $\rho(t)$ with the Bell state after this rotation. 

We will plug in different values for $\Gamma$, to explore under which conditions of $J/\Gamma$ the qubits become entangled.

```python
# values of J/Gamma to try
Gamma_min = 0
Gamma_max = 2 * J
pts = 201
Gammas = np.linspace(Gamma_min, Gamma_max, pts)

results = []

for Gamma in Gammas:
    results.append(solve_me(J, Gamma, [0, np.pi/(4*J)]))
    
# apply the single qubit rotation along X of qubit A
U = tensor(identity(2)*np.cos(np.pi/4) - np.complex(0, 1)*sigmax()*np.sin(-np.pi/4), identity(2))
U_inv = tensor(identity(2)*np.cos(np.pi/4) + np.complex(0, 1)*sigmax()*np.sin(-np.pi/4), identity(2))
for count, result in enumerate(results): 
    results[count] = U*result*U_inv
```

```python
# Calculate the entanglement fidelity

psi_plus = (tensor(basis(2,1), basis(2,0)) + tensor(basis(2,0), basis(2,1))).unit()
rho_psi_plus = ket2dm(psi_plus)
fidelities = []

for result in results: 
    fidelities.append(fidelity(result[1], rho_psi_plus)**2)
    
fidelities = np.array(fidelities)
```

```python
# plot the result

plt.plot(Gammas / J, fidelities)
plt.xlabel('$\Gamma$ (units of $J$)')
plt.ylabel('Entanglement fidelity')
```

```python
# 3.c.i: find numerically the ratio J/Gamma requried to create entangled states

# Recall that the list 'Gammas' contains the Gamma values, 
# and 'fidelities' contain the entanglement fidelities at time t = pi/(4J).
min_ratio = J / np.max(Gammas[np.where(fidelities > 1 / 2)[0]])
print(f'The minimum ratio J/Gamma needed for entanglement is {min_ratio}')
```

```python
# 3.c.iii: calculate the entanglement fidelity for Gamma = J/100
# recall that 'fidelities' contain the entanglement fidelities at time t = pi/(4J).
Gamma = J / 100
result = solve_me(J, Gamma, [0, np.pi/(4*J)])
    
# apply the single qubit rotation along X of qubit A
U = tensor(identity(2)*np.cos(np.pi/4) - np.complex(0, 1)*sigmax()*np.sin(-np.pi/4), identity(2))
U_inv = tensor(identity(2)*np.cos(np.pi/4) + np.complex(0, 1)*sigmax()*np.sin(-np.pi/4), identity(2))
result = U*result*U_inv
    
# Calculate the entanglement fidelity
psi_plus = (tensor(basis(2,1), basis(2,0)) + tensor(basis(2,0), basis(2,1))).unit()
rho_psi_plus = ket2dm(psi_plus)
fidelity(result[1], rho_psi_plus)**2
```
