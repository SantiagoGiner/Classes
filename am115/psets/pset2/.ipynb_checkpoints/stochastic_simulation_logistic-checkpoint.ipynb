Simulation of the stochastic logistic equations, based on "Modelling biological populations in space and time", Eric Renshaw

```python
# Set parameters for logistic equation: X_t=R*X*(1-X/K)
R = 1.0
K = 50
dimP = K * 10  # number of points at which pdf is defined, need to be larger than carrying capacity:

# Corresponding birth-death probability, see lecture notes for the mapping
mu = R / K
lam = R * (1 - (1 / K))
```

```python
# Simulate the stochastic logistic equation, using section 3.4
nsteps = 2500
N = np.zeros(nsteps)
N[0] = 3
t = 0
dt = 10**(-3)

# Think about what happens if dt is too large, and why?
while t < nsteps - 1 and N[t] > 0:
    r = np.random.rand(1)
    # Birth happens
    if r < lam * N[t] * dt:
        N[t + 1] = N[t] + 1
    # Death happens
    elif r < (lam * N[t] + mu * N[t] * (N[t] - 1)) * dt:
        N[t + 1] = N[t] - 1
    # No birth or death
    else:
        N[t + 1] = N[t]
    t += 1
```

```python
# Plot solution
fig, ax = plt.subplots(1,1,figsize=(9,7))

ax.plot(np.arange(nsteps), N, linewidth=1.5)

# Set title and axes labels
ax.set_title('stochastic simulation')
ax.set_xlabel('time steps')
ax.set_ylabel(r'$N(T)$')

plt.show()
```

## Plotting the Distribution over Time

```python
# Set the number of repetitions and steps / repetition
n_reps = 5000
n_steps = 10001
step_size = 100

# Array to store population numbers at every time step
pops = np.zeros((int(n_steps / step_size), n_reps))

# Iterate over number of repetitions
for rep in range(n_reps):
    # Simulate the stochastic logistic equation, using section 3.4
    N = np.zeros(n_steps)
    N[0] = 3
    t = 0
    dt = 10**(-3)

    # Think about what happens if dt is too large, and why?
    while t < n_steps - 1 and N[t] > 0:
        r = np.random.rand(1)
        # Birth happens
        if r < lam * N[t] * dt:
            N[t + 1] = N[t] + 1
        # Death happens
        elif r < (lam * N[t] + mu * N[t] * (N[t] - 1)) * dt:
            N[t + 1] = N[t] - 1
        # No birth or death
        else:
            N[t + 1] = N[t]
        t += 1
        if t % 100 == 0:
            pops[int(t / step_size - 1)][rep] = N[t]
        
# Plotting
fig = plt.figure(figsize=(40, 40))
for i in range(10):
    for j in range(10):
        ax = plt.subplot2grid((10, 10), (i, j))
        ax.hist(pops[10 * i + j])
```

```python
plot_times = np.arange(1, 10) * 10
fig = plt.figure(figsize=(40, 40))
for i in range(3):
    for j in range(3):
        ax = plt.subplot2grid((3, 3), (i, j))
        ax.hist(pops[plot_times[3 * i + j]])
```

Let us compare what the distribution of the population looks like at the end of 10,000 time steps and the Truncated Poisson distribution, to which the distribution approaches.

```python
from scipy.special import factorial

ratio = lam / mu
N_arr = np.linspace(10, 90, 45)
tpois = np.divide(np.divide(np.power(ratio, N_arr), factorial(N_arr)), np.expm1(ratio))

fig, ax = plt.subplots(1, 1, figsize=(10, 8))
ax.hist(pops[-1], density=True, bins=90, label="Simulation Data")
# Plot Truncated Poisson
ax.plot(N_arr, tpois / np.sum(tpois), label="Analytic Result", color="crimson", linewidth=3)
ax.set_xlabel("$n$")
ax.set_ylabel("$P(N = n)$")
ax.set_title("Probability Distribution of $N$ after 10,000 Time Steps", fontsize=20)
ax.legend()
plt.show()
```
