## Problem 2
For at least one of the dice to be 1, either one of them is a 1 and the other is any other number, or both of them are 1. The former of these occurs with probability $2 \cdot 1/6 \cdot 5/6 = 5/18,$ since the probability that one die shows 1 is $1/6$ and the probability that the other die shows any other number is $5/ 6,$ and there are two possiblities for which die shows which number. Finally, the probability that both dice are 1 is $1/6 \cdot 1/6 = 1/36.$ Therefore

$$\boxed{P(\text{at least one die shows 1}) = \frac{5}{18} + \frac{1}{36} = \frac{11}{36} = 0.30\bar 5.}$$

Let us verify by simulation...

```python
# Verifying by simulation
def at_least_one():
    die1, die2 = np.random.randint(1, 6 + 1), np.random.randint(1, 6 + 1)
    if die1 == 1 or die2 == 1:
        return True  

n_sim = int(10**5)
n_occur = 0
for i in range(n_sim):
    if at_least_one():
        n_occur += 1
print(f'Experimental probability: {n_occur / n_sim}')
```

We do get a result very close to the analytical result!

## Problem 5

We know that the probability of die A showing a 1 is simply $1/6.$ Let us verify that the Law of Total Probability (LOTP) holds by rederiving this result conditioning on the outcome of die B. The LOTP says

$$P(\text{die A shows 1}) = \sum_{i=1}^6 P(\text{die A shows 1}| \text{die B shows } i) \cdot P(\text{die B shows } i),$$

where $P(\text{die A shows 1}| \text{die B shows } i) = 1/6$ since the die outcomes are independent and $P(\text{die B shows } i) = 1/6.$ Thus, we have

$$P(\text{die A shows 1}) = \sum_{i=1}^6 \frac16 \cdot\frac16 = \frac{6}{36} = \boxed{\frac16,}$$

as desired.

```python
# Define number of repetitions
n_reps = int(10**4)

# Array in which to save number of occurences of 1 for die A for each of B's six outcomes
dieA_one_counts = np.zeros(6, dtype=int)

# Array to count number of occurences for die B of each of 1, ..., 6
dieB_counts = np.zeros(6, dtype=int)

# Simulate n_reps number of trials
for i in range(n_reps):
    dieB = np.random.randint(1, 6 + 1)
    dieB_counts[dieB - 1] += 1
    # Add to count of 1 for die A at the index of B's outcome
    dieA = np.random.randint(1, 6 + 1)
    if dieA == 1:
        dieA_one_counts[dieB - 1] += 1
        
# Get prediction from LOTP and print it
conditional_probs = dieA_one_counts / dieB_counts 
lotp_pred = np.sum([cond * 1 / 6 for cond in conditional_probs])
print(f"Experimental probability: {lotp_pred}")
```

Which is indeed very close to $1/6 = 0.1\overline 6.$

By symmetry, all outcomes are equally likely, and we can see that there are exactly 2 out of the 11 outcomes where one die shows a 6 and the other a one, so the desired probability is $2/11,$

$$\boxed{P(\text{other die shows 1} | \text{at least one shows 6}) = \frac{2}{11} = 0.\overline{18}.}$$

We again verify this below via simulation...

```python
# Number of trials to simulate
n_reps = int(10**5)

# Initialize variables and iterate
num, denom = 0, 0
for i in range(n_reps):
    die1, die2 = np.random.randint(1, 6 + 1), np.random.randint(1, 6 + 1)
    # Whenever a die is 6, this is one of the total outcomes we are considering
    if die1 == 6 or die2 == 6:
        denom += 1
        # If one of the dice is 6 and the other 1, this is a desired outcome
        if die1 == 1 or die2 == 1:
            num += 1
print(num / denom)
```

## Problem 7

### Part a

A pooled sample will test positive if at least one of the people in the pool is infected. Therefore, we are looking to find the probability that at least one of the people in a pool size of $s$ is infected. It is easier to look at the complement, the probability that *no one* in the group is infected, and take 1 minus this result to obtain the desired probability. Any one person is not infected with probability $1 - p,$ and since we are assuming independence between the test status of different people, the whole pool of $s$ will have no infected people with probability $(1 - p)^s,$ and therefore,

$$\boxed{P(\text{pool tests positive}) = 1 - (1-p)^s.}$$

### Part b

We can write $T$ as a sum of random variables (r.v.s.), specifically, we can write it as the sum of the number of tests given to each pool. Let $T_j$ be the r.v. counting the number of tests given to pool $j,$ for $j \in {1, 2, \dots, N/s},$ since there are a total of $N/s$ pools. By linearity and symmetry, that is, since the expected number of tests for any one pool must be the same for all pools, and in particular to that of, say, pool 1, we then have that

$$E(T) = \frac{N}{s} E(T_1).$$

Now we have to find the expected number of tests given to any particular pool. This can be done using the Law of Total Expectation (LOTE), which is the analogue of LOTP but with expectation instead of probability. As we have seen, with probability $1 - (1-p)^s,$ at least one person in the pool is infected, and we would therefore have to perform $s + 1$ tests: one for every person in the pool and the initial test. And with probability $1 - [1 - (1-p)^s] = (1-p)^s,$ no one in the pool is infected, and therefore we only have to do $1$ test. Therefore, we obtain that

$$E(T_1) = [1 - (1-p)^s] \cdot (s + 1) + (1-p)^s \cdot 1 = 1 + s(1 - (1 - p))^s,$$

and thus

$$\boxed{E(T) = N \left(\frac{1}{s} + 1 - (1-p)^s\right).}$$

### Part c

Let us first plot this expected value to get a sense of what it looks like, with respect to $s,$ for different values of $p.$

```python
exp_tests = lambda s, p, N: N * ((1 / s) + 1 - (1 - p)**s)

# Define constants
N = 100
s_axis = np.linspace(1, N, 400)
p_arr = np.linspace(0.001, 0.4, 10)

fig, ax = plt.subplots(1, 1, figsize=(12, 8))
for p in p_arr:
    ax.plot(s_axis, exp_tests(s_axis, p, N), label=f"$p = {p:.4f}$")
ax.axhline(100, linestyle="--", color="k", label=r"$E(T) = N$")

ax.set_xlabel(r"$s$")
ax.set_ylabel(r"$E(T)$")
ax.set_title(f"$N = {N}$", fontsize=22)
ax.legend()
plt.show()
```

These plots make sense: as $p$ increases, we expect to have to perform more tests, since more and more people have the disease, so the probability that any pool tests positive is very high, and in the end we expect that as the pool sizes get larger, the number of tests approaches $N,$ the population size, since in the limit that $N = s$ and $p$ is reasonably high, then the probability that the person in the pool we test, which is one person in the whole population, tests positive is quite high, in which case, we would have to test everyone in the population and so $E(T) \to N.$

Let us now minimize this function...

```python
# Use the fsolve root-finding function from SciPy
from scipy.optimize import fsolve

# Define derivative of E(T) as a function of s, p
dE_ds = lambda s, p: -1 / s**2 - np.log(1 - p) * (1 - p)**s

# Define constants
p_arr = np.linspace(0.001, 0.4, 200)

# Array in which to store optimal values of s
s_optimal = np.zeros(200)

# Iterate over each p
for i, p in enumerate(p_arr):
    s_optimal[i] = fsolve(dE_ds, x0=0.1, args=(p))[0]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
ax.plot(p_arr, s_optimal)
ax.axhline(1, linestyle="--", color="grey", label="s = 1")

ax.set_xlabel(r"$p$")
ax.set_ylabel(r"$s_{\rm{optimal}}$")
ax.set_title(f"Optimal Value of $s$", fontsize=22)
ax.legend()
plt.show()
```

## Problem 8

### Part 1

```python
# Define constants
sigma = 5.67e-8  # Stephan-Boltzmann constant
H = 1.0e8  # heat capacity of 10^8 Joules/Kelvin

# Define a function to the temperature tendency of the planet
def dT_dt(T, S0):
    """ 
    Adapted from provided code 
    Compute the temperature tendency of the planet (in Kelvin/seconds)
    given global mean surface temperature T (in Kelvin) and the solar 
    constant S0 (in Watts per meter^2)

    Args:
        T (float): global mean surface temperature range (in Kelvin)
        S0 (float): solar constant (in Watts per meter^2)

    Returns:
        float: temperature tendency of the planet (in Kelvin/seconds)
    """
    # Define the albedo and find and return T
    albedo = max(min(0.6, 0.3 * (1 + (273 - T) / 10)), 0.3)
    return ((1 - albedo) * S0 / 4 - sigma * T**4) / H
```

```python
# Defining constants
S0 = 2200  # W/m^2
T_range = np.linspace(220, 300, 200)

# Compute dT/dt values for each T in T_range
dT_dt_arr = np.array([dT_dt(T, S0) for T in T_range])

fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ax.plot(T_range, dT_dt_arr)
ax.axhline(0, linestyle="--", color="gray", alpha=0.8)
ax.set_xlabel(r"$T$ [K]", fontsize=16)
ax.set_ylabel(r"$dT/dt$ [K/s]", fontsize=16)
plt.show()
```

### Part 2

The critical points satisfy $dT/dt = 0.$ We will solve this equation for every interval that defined the albedo $\alpha(T).$

<ul>
    <li> $T > 273$: $T_1 = \left(\frac{0.4 S_0}{4\sigma}\right)^{1/4} \approx \boxed{249.6 \text{ K}.}$ </li>
    <li> 
        $263 \leq T \leq 273$: Here we get the equation
        $$-\frac{7.19 S_0}{4} + \frac{0.03 S_0}{4} T - \sigma T^4 = 0,$$
        which we solve numerically below to obtain $\boxed{T_2 \approx 267.2 \text{ K}.}$
    </li>
    <li> $T > 273$: $T_2 = \left(\frac{0.7 S_0}{4\sigma}\right)^{1/4} \approx \boxed{287.1 \text{ K}.}$ </li>
</ul>

```python
from scipy.optimize import root

f = lambda t: -7.49 * S0 / 4 + (0.03 * S0 / 4) * t - sigma * t**4

t1 = ((1 - 0.6) * S0 / (4 * sigma))**(1/4)
t2 = root(f, 270).x[0]
t3 = ((1 - 0.3) * S0 / (4 * sigma))**(1/4)

print(f"First critical point: {t1}")
print(f"Second critical point: {t2}")
print(f"Third critical point: {t3}")
```

These are shown in the plot below.

```python
# Defining constants
S0 = 2200  # W/m^2
T_range = np.linspace(220, 300, 200)

# Compute dT/dt values for each T in T_range
dT_dt_arr = np.array([dT_dt(T, S0) for T in T_range])

fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ax.plot(T_range, dT_dt_arr)
ax.axhline(0, linestyle="--", color="gray", alpha=0.8, label="dT/dt = 0")
ax.plot(t1, 0, marker="o", markersize=10, color="green", label="Stable Critical Point")
ax.plot(t2, 0, marker="o", markersize=10, color="red", label="Unstable Critical Point")
ax.plot(t3, 0, marker="o", markersize=10, color="green")
ax.set_xlabel(r"$T$ [K]", fontsize=16)
ax.set_ylabel(r"$dT/dt$ [K/s]", fontsize=16)
ax.legend()
plt.show()
```

The points $T \approx 249.6 \text{ K}$  and $287.1 \text{ K}$ are both stable critical points since the time derivative of $T$ goes from being positive to the left of the point and being positive to the right, meaning that temperatures to the left of each green point tend to increase and move right, closer to the point, and temperatures to the right and therefore higher in temperature than the green point, tend to move to the left and decrease back to the critical point. By the exact same reasoning, the point in the middle, namely, $T \approx 267.2 \text{ K}$ is unstable, since temperatures to its left tend to move further to the left, since the derivative is negative there, and points to its right tend to move further to the right, since the derivative is positive there.

### Part 3

The warm stable fixed point is the point $T \approx 287.1 \text{ K}$ since it is the only stable point above freezing.

```python
# Define constants
n_days = 30000
n_steps = 1000
dt = n_days / n_steps 

# Determining step size of S0
S0_start, S0_end = S0, 0.7 * S0
dS0 = (S0_end - S0_start) / n_steps

# Array to store temperatures
T_arr_desc = np.zeros(n_steps)
T_arr_desc[0] = t3  # start at the warm stable point

# Array to store descending values of S0
S0_arr_desc = np.zeros(n_steps)
S0_arr_desc[0] = S0_start

# Array for times
times = np.zeros(n_steps)

# Iterate over each day
for i in range(n_steps - 1):
    # Advance time
    times[i + 1] = times[i] + dt
    # Get S0 and slope and convert slope to K / day
    S0_arr_desc[i + 1] = S0_arr_desc[i] + dS0
    slope_day = dT_dt(T_arr_desc[i], S0_arr_desc[i]) * 3600 * 24
    # Advance using Euler's method
    T_arr_desc[i + 1] = T_arr_desc[i] + slope_day * dt
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ax.plot(S0_arr_desc, T_arr_desc)
ax.invert_xaxis()

ax.set_xlabel(r"$S_0$ [W / $\rm{m}^2$]", fontsize=16)
ax.set_ylabel(r"$T$ [K]", fontsize=16)
plt.show()
```

```python
# Define constants
n_days = 60000
n_steps = 2000
dt = n_days / n_steps 

# Determining step size of S0
S0_start, S0_end = S0_arr_desc[-1], 1.3 * S0
dS0 = (S0_end - S0_start) / n_steps

# Array to store temperatures
T_arr_asc = np.zeros(n_steps)
T_arr_asc[0] = T_arr_desc[-1]

# Array to store ascending values of S0
S0_arr_asc = np.zeros(n_steps)
S0_arr_asc[0] = S0_start

# Array for times
times = np.zeros(n_steps)

# Iterate over each day
for i in range(n_steps - 1):
    # Advance time
    times[i + 1] = times[i] + dt
    # Get S0 and slope and convert slope to K / day
    S0_arr_asc[i + 1] = S0_arr_asc[i] + dS0
    slope_day = dT_dt(T_arr_asc[i], S0_arr_asc[i]) * 3600 * 24
    # Advance using Euler's method
    T_arr_asc[i + 1] = T_arr_asc[i] + slope_day * dt
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ax.plot(S0_arr_asc, T_arr_asc)

ax.set_xlabel(r"$S_0$ [W / $\rm{m}^2$]", fontsize=16)
ax.set_ylabel(r"$T$ [K]", fontsize=16)
plt.show()
```

Below we plot both temperature trends.

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ax.plot(S0_arr_desc, T_arr_desc, label="Decreasing $S_0$")
ax.plot(S0_arr_asc, T_arr_asc, label="Increasing $S_0$")

ax.set_xlabel(r"$S_0$ [W / $\rm{m}^2$]", fontsize=16)
ax.set_ylabel(r"$T$ [K]", fontsize=16)
ax.legend()
plt.show()
```

These results make sense. First, we see that as we decrease $S_0,$ the Earth's average surface temperature increases, as expected, since we are decreasing the power per surface area transmitted by the Sun. Similarly, as we increase $S_0,$ we see how the temperature increases. We also see sudden jumps in how the temperature changes over time, which correspond to bifurcations in the dynamics of the system's evolution, as we saw in class in the context of modeling a species' evolution in time with ODEs. What we are essentially plotting above are the critical points of the climate model, and the sudden drop or spike are bifurcation points. This is another demonstration of hysteresis, that is, that a system's evolution in time at any given point in time depends not just on the current conditions but on it history, i.e. how it got to that point. When decreasing $S_0,$ we see a sudden drop at $T \sim 1800 \text{ K},$ but when we are increasing $S_0,$ the temperature inceases gradually through the $T \sim 1800 \text{ K}$ and has a sudden spike at around $T \sim 2800 \text{ K}.$

## Plotting the Distribution over Time

```python
# Set parameters for logistic equation: X_t=R*X*(1-X/K)
R = 1.0
K = 50
dimP = K * 10  # number of points at which pdf is defined, need to be larger than carrying capacity:

# Corresponding birth-death probability, see lecture notes for the mapping
mu = R / K
lam = R * (1 - (1 / K))
```

```python
# Set the number of repetitions and steps / repetition
n_reps = 5000
n_steps = 10001
step_size = 100

# Array to store population numbers at every time step
pops = np.zeros((int(n_steps / step_size), n_reps))

# Iterate over number of repetitions
for rep in range(n_reps):
    # Simulate the stochastic logistic equation, using section 3.4
    N = np.zeros(n_steps)
    N[0] = 3
    t = 0
    dt = 10**(-3)

    # Think about what happens if dt is too large, and why?
    while t < n_steps - 1 and N[t] > 0:
        r = np.random.rand(1)
        # Birth happens
        if r < lam * N[t] * dt:
            N[t + 1] = N[t] + 1
        # Death happens
        elif r < (lam * N[t] + mu * N[t] * (N[t] - 1)) * dt:
            N[t + 1] = N[t] - 1
        # No birth or death
        else:
            N[t + 1] = N[t]
        t += 1
        if t % 100 == 0:
            pops[int(t / step_size - 1)][rep] = N[t]
        
# Plotting
fig = plt.figure(figsize=(40, 40))
for i in range(10):
    for j in range(10):
        ax = plt.subplot2grid((10, 10), (i, j))
        ax.hist(pops[10 * i + j])
```

```python
plot_times = [500, 1500, 3000, 6000, 8000, 10000]
k = 0
fig = plt.figure(figsize=(14, 24))
for i in range(3):
    for j in range(2):
        ax = plt.subplot2grid((3, 2), (i, j))
        # The first plot needs coarser binning, as most population numbers are the same across repetitions
        if k == 0:
            ax.hist(pops[int(plot_times[k] / step_size) - 1], density=True, bins=10)
        else:
            ax.hist(pops[int(plot_times[k] / step_size) - 1], density=True, bins=30)
        ax.set_xlabel("$n$")
        ax.set_ylabel("$P(N=n)$")
        ax.set_title(f"Probability Distribution after {plot_times[k]} Steps")
        k += 1
plt.show()
```

Let us compare what the distribution of the population looks like at the end of 10,000 time steps and the Truncated Poisson distribution, to which the distribution approaches.

```python
from scipy.special import factorial

ratio = lam / mu
N_arr = np.linspace(10, 90, 45)
tpois = np.divide(np.divide(np.power(ratio, N_arr), factorial(N_arr)), np.expm1(ratio))

fig, ax = plt.subplots(1, 1, figsize=(10, 8))
ax.hist(pops[-1], density=True, bins=90, label="Simulation Data")
# Plot Truncated Poisson
ax.plot(N_arr, tpois / np.sum(tpois), label="Analytic Result", color="crimson", linewidth=3)
ax.set_xlabel("$n$")
ax.set_ylabel("$P(N = n)$")
ax.set_title("Probability Distribution of $N$ after 10,000 Time Steps", fontsize=20)
ax.legend()
plt.show()
```

As the population starts at a small value of 3, relative to the carrying capacity of 50, initially we see the mean of the probability distribution grow over time. The population continues to grow while it is significantly less than carrying capacity. This is shown in the 500 timestep plot where the mean looks to be about 5. After 1000 more timesteps, at timestep 1500, we see the distribution has evolved to have a mean of roughly 12. This evolution continues. After 3000 timesteps the distribution looks to have a mean of roughly 25, and after 6000 timesteps, a mean of roughly 48. After 8000 timesteps, however, something interesting happens; the probability distribution of the population have evolved to have a mean of K = 50, reaching the carrying capacity. Thus, even after 2000 more timesteps, at timestep 10000, the probabilty distribution looks roughly identical, with a mean still at roughly 50. This is because after a large number of time steps the population reaches its statistically steady state, where, despite small fluctuations in time, the probability distribution (defined as the fraction of time that the population spends at each value) remains constant. As seen by the plot above, the steady-state distribution has a peak at the carrying capacity K = 50 but also a significant spread of values at which the population spends time. It also aligns with the truncated poisson distribution with the same parameters, the analytical solution for our steady-state probability distribution as shown in the class notes.

The main factors that would affect our decision of how many roses to buy would be:
<ul>
    <li>
        Supply related:
        <ul>
            <li> Price of a rose: this factor could be obtained by surveying local providers and inquiring about their prices
            <li> Availability of roses: we could again inquire each provider in the local area about the number of roses that they would be able to supply to us </li>
            <li> Quality of the roses selected: this factor is the most subjective of the ones considered, but we could take an approach where we rate each rose that we get from a provider on a scale from 1 through 10 and use that number to determine the scaling  </li> 
        </ul>
    </li>
    <li>
        Demand related:
        <ul>
            <li> The number of couples in a relationship in the area that I live in: obtaining this number via a survey would be complicated, as there could be many residents in the area we plan on selling the roses, but rather than surveying every single person in the area, we could conduct a randomized survey where we sample only a small portion of the peopla in the area and ask them about their relationship status. The most naive approach, i.e. the simplest model, would then consist of using the percentage of people in a relationship we find in the survey and multiply that by the total population. We could make this more complicated, as there are whole areas of study dedicated to estimating a random variable from different sampling methods, but we think this approach is justified as a first tackle at the problem.</li>
            <li> The number of couples that give roses as gifts: we would follow a similar approach to the previous factor, where we would estimate this quantity by sampling a random portion of the population </li>
            <li> Number of roses a person gives per gift: we could again estimate this via a survey, asking the people in the survey who affirm they buy roses about the number of roses they usually gift </li>
        </ul>
    </li>
</ul>

Once we have the data for the supply-related factors, we could build an objective function $L(\text{price}, \text{availability}, \text{quality})$ that assigns some weight to each quantity, depending on how much we value price vs. availability vs. quality (which is something that will vary from vendor to vendor), and then evaluate each rose provider that we know based on this objective function and choose the provider that maximizes it (or minimizes it, depending on the specifics of the function).

Once we have selected a specific provider and have also collected the demand-related data, we could model the number of customers that we will get in our store as a Poisson distribution. The mean of this distribution, i.e. the expected number of customers, would be the total population in the area times the percentage of people in a relationship times the percentage of people in a relationship who gift roses. That is, if $N$ is the number of customers we get, we would model $N$ as

$$N \sim \text{Pois}(np_1p_2),$$

where $p_1, p_2$ are our estimates for the fraction of people in a relationship and who gift roses, respectively.

The Poisson distribution could be an appropriate approximation in this case because the probability that any one person buys a rose is low, but the population is assumed to be large, so on average, we expect a fairly large number of people to buy a rose. Finally, to obtain the number of roses to order, we would multiply the expected value of $N$ times the average number of roses each person buys, as obtained from the survey.

Some potential drawbacks of this model are:

<ul>
    <li> It does not take into account people in the population who are not in a relationship but would still gift a rose, e.g. to a potential lover or friend </li>
    <li> The Poisson distribution assumes all events, i.e. someone buying a rose, are independent of each other, when in reality if someone buys a rose, it might make another person want to do the same, and therefore there is some dependence between events </li>
    <li> It assumes every person will give the same number of rosesw when in reality different people will buy a different number </li>
</ul>
