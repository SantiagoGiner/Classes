```python
import numpy as np
import matplotlib.pyplot as plt
```

### Part 1

```python
# Define constants
sigma = 5.67e-8  # Stephan-Boltzmann constant
H = 1.0e8  # heat capacity of 10^8 Joules/Kelvin

# Define a function to the temperature tendency of the planet
def dT_dt(T, S0):
    """ 
    Adapted from provided code 
    Compute the temperature tendency of the planet (in Kelvin/seconds)
    given global mean surface temperature T (in Kelvin) and the solar 
    constant S0 (in Watts per meter^2)

    Args:
        T (float): global mean surface temperature range (in Kelvin)
        S0 (float): solar constant (in Watts per meter^2)

    Returns:
        float: temperature tendency of the planet (in Kelvin/seconds)
    """
    # Define the albedo and find and return T
    albedo = max(min(0.6, 0.3 * (1 + (273 - T) / 10)), 0.3)
    return ((1 - albedo) * S0 / 4 - sigma * T**4) / H
```

```python
# Defining constants
S0 = 2200  # W/m^2
T_range = np.linspace(220, 300, 200)

# Compute dT/dt values for each T in T_range
dT_dt_arr = np.array([dT_dt(T, S0) for T in T_range])

fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ax.plot(T_range, dT_dt_arr)
ax.axhline(0, linestyle="--", color="gray", alpha=0.8)
ax.set_xlabel(r"$T$ [K]", fontsize=16)
ax.set_ylabel(r"$dT/dt$ [K/s]", fontsize=16)
plt.show()
```

### Part 2

The critical points satisfy $dT/dt = 0.$ We will solve this equation for every interval that defined the albedo $\alpha(T).$

<ul>
    <li> $T > 273$: $T_1 = \left(\frac{0.4 S_0}{4\sigma}\right)^{1/4} \approx \boxed{249.6 \text{ K}.}$ </li>
    <li> 
        $263 \leq T \leq 273$: Here we get the equation
        $$-\frac{7.19 S_0}{4} + \frac{0.03 S_0}{4} T - \sigma T^4 = 0,$$
        which we solve numerically below to obtain $\boxed{T_2 \approx 267.2 \text{ K}.}$
    </li>
    <li> $T > 273$: $T_2 = \left(\frac{0.7 S_0}{4\sigma}\right)^{1/4} \approx \boxed{287.1 \text{ K}.}$ </li>
</ul>

```python
from scipy.optimize import root

f = lambda t: -7.49 * S0 / 4 + (0.03 * S0 / 4) * t - sigma * t**4

t1 = ((1 - 0.6) * S0 / (4 * sigma))**(1/4)
t2 = root(f, 270).x[0]
t3 = ((1 - 0.3) * S0 / (4 * sigma))**(1/4)

print(f"First critical point: {t1}")
print(f"Second critical point: {t2}")
print(f"Third critical point: {t3}")
```

These are shown in the plot below.

```python
# Defining constants
S0 = 2200  # W/m^2
T_range = np.linspace(220, 300, 200)

# Compute dT/dt values for each T in T_range
dT_dt_arr = np.array([dT_dt(T, S0) for T in T_range])

fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ax.plot(T_range, dT_dt_arr)
ax.axhline(0, linestyle="--", color="gray", alpha=0.8, label="dT/dt = 0")
ax.plot(t1, 0, marker="o", markersize=10, color="green", label="Stable Critical Point")
ax.plot(t2, 0, marker="o", markersize=10, color="red", label="Unstable Critical Point")
ax.plot(t3, 0, marker="o", markersize=10, color="green")
ax.set_xlabel(r"$T$ [K]", fontsize=16)
ax.set_ylabel(r"$dT/dt$ [K/s]", fontsize=16)
ax.legend()
plt.show()
```

The points $T \approx 249.6 \text{ K}$  and $287.1 \text{ K}$ are both stable critical points since the time derivative of $T$ goes from being positive to the left of the point and being positive to the right, meaning that temperatures to the left of each green point tend to increase and move right, closer to the point, and temperatures to the right and therefore higher in temperature than the green point, tend to move to the left and decrease back to the critical point. By the exact same reasoning, the point in the middle, namely, $T \approx 267.2 \text{ K}$ is unstable, since temperatures to its left tend to move further to the left, since the derivative is negative there, and points to its right tend to move further to the right, since the derivative is positive there.

### Part 3
