```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Define a function to compute analytical solution to the logistic
# equation at different growth rates and carrying capacities
def y(t, N0, r=0.5, K=10.):
    """Calculate the analytical solution to the logistic equation

    Args:
        t (numpy.ndarray): time
        N0 (numpy.float64): initial population (unit: # of animals)
        r (float): growth rate (unit 1/time)
        K (float): carrying capacity (unit: # of animals)

    Returns:
        numpy.ndarray: analytical solution to the logistic equation
    """
    
    N = K / (1. - ((N0-K)/N0) * np.exp(-r*t) )
    
    return N 
```

```python
# Define variables (only vary the initial conditions)
t = np.linspace(0,10,101)              # 101 points in time spanning [0, 10]
arr_N0 = 2. * np.arange(1,11)          # array of 10 initial conditions 
r = 0.5                                # growth rate
K = 10.                                # carrying capacity
```

```python
# Customize for matplotlib
# If interested in the matplotlib object hierarchy, check: https://realpython.com/python-matplotlib-guide/
plt.rcParams['axes.linewidth'] = 1
plt.rcParams['xtick.bottom'] = True
plt.rcParams['ytick.left'] = True
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['mathtext.default'] = 'regular'
# Change font size: http://www.futurile.net/2016/02/27/matplotlib-beautiful-plots-with-style/
plt.rcParams['font.size'] = 12
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['axes.labelweight'] = 'bold'
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 14
plt.rcParams['figure.titlesize'] = 20
```

```python
# Plot solutions for different initial population choices
fig, ax = plt.subplots(1,1,figsize=(9,7))

# Create a color gradient for plotting all solutions
# Check out more colormaps: https://matplotlib.org/stable/tutorials/colors/colormaps.html
n = len(arr_N0) # number of solutions
colors = plt.cm.coolwarm(np.linspace(0.,1.,n))

# Loop over initial conditions and plot each solution
for idx, n0 in enumerate(arr_N0):
    ax.plot(t, y(t, n0, r=r, K=K), linewidth=1.5, color=colors[idx])

# Plot the carrying capacity as a reference
# zorder set the line to be at the bottom of this figure
ax.plot([t.min(), t.max()], [K, K], color='k', linewidth=4.5, label='Carrying Capacity', zorder=0)   

# Set axes range
ax.set_xlim(0,10)
ax.set_ylim(0,20)

# Set axes labels and figure title
ax.set_xlabel('time')
ax.set_ylabel('N')
ax.set_title(f'K = {K}, r = {r}') 
ax.legend(loc='upper right')

plt.show()
```

# Varying the Growth Rate

```python
# Set value of the parameters, creating an array of different growth rates
t = np.linspace(0, 10, 101)
N_0 = 2
r_arr = np.linspace(0.1, 1, 10)
K = 10
```

```python
# Plot solutions for different initial population choices
fig2, ax2 = plt.subplots(1, 1, figsize=(9,7))

# Create a color gradient for plotting all solutions
# Check out more colormaps: https://matplotlib.org/stable/tutorials/colors/colormaps.html
colors = plt.cm.coolwarm(np.linspace(0, 1, len(r_arr)))

# Loop over growth rates and plot each solution
for idx, r in enumerate(r_arr):
    ax2.plot(t, y(t, N_0, r=r, K=K), linewidth=1.5, color=colors[idx])

# Plot the carrying capacity as a reference
# zorder sets the line to be at the bottom of this figure
ax2.plot([t.min(), t.max()], [K, K], color='k', linewidth=4.5, label='Carrying Capacity', zorder=0)   

# Set axes range
ax2.set_xlim(0, 10)
ax2.set_ylim(-1, 20)

# Set axes labels and figure title
ax2.set_xlabel('time')
ax2.set_ylabel('Growth Rate')
ax2.set_title(f'$N_0$ = {N_0}, $K$ = {K}') 
ax2.legend(loc='upper right')

fig2.savefig('./growth_rate.pdf', bbox_inches='tight')

plt.show()
```

# Varying the Carrying Capacity

```python
# Set value of the parameters, creating an array of different growth rates
t = np.linspace(0, 10, 101)
N_0 = 2
r = 0.5
K_arr = np.arange(8, 18)
```

```python
# Plot solutions for different initial population choices
fig3, ax3 = plt.subplots(1, 1, figsize=(9,7))

# Create a color gradient for plotting all solutions
# Check out more colormaps: https://matplotlib.org/stable/tutorials/colors/colormaps.html
colors = plt.cm.coolwarm(np.linspace(0, 1, len(K_arr)))

# Loop over growth rates and plot each solution
for i, K in enumerate(K_arr):
    ax3.plot(t, y(t, N_0, r=r, K=K), linewidth=1.5, color=colors[i])
    ax3.plot([t.min(), t.max()], [K, K], '--', color=colors[i], linewidth=1.0, alpha=1.0)  

# Set axes range
ax3.set_xlim(0,10)
ax3.set_ylim(0,20)

# Set axes labels and figure title
ax3.set_xlabel('time')
ax3.set_ylabel('Carrying Capacity')
ax3.set_title(f'$N_0$ = {N_0}, $r$ = {r}') 

fig3.savefig('carrying_capacity.pdf', bbox_inches='tight')

plt.show()
```

Above we have included a dashed line for each of the values of the carrying capacity to guide the eye as to where the carrying capacity lies for each of the different graphs. The plots with some carrying capacity $K'$ have the same color as the dashed line for $K'.$
