```python
import numpy as np
import matplotlib.pylab as plt
from scipy.integrate import ode
```

```python
# Creating a function equivalent to ode45 in Matlab
def ode45(f, tint, y0, rtol=1e-9):
    r = ode(f).set_integrator('zvode', method='bdf', rtol=rtol)
    r.set_initial_value(y0, tint[0])
    dt = (tint[1] - tint[0]) / 10000
    t1 = tint[1]
    tt = []
    rr = []
    while r.successful() and r.t < t1:
        tt.append(r.t + dt)
        rr.append(np.real(r.integrate(r.t+dt)))
    return tt, rr
```

```python
# An example illustrating the definition 
# of a function for use in ode45
def ode_function():
    y0 = 1;
    r = -1;
    [t,y] = ode45(lambda t, y: f(t, y, r),[0,5], y0)
    plt.plot(t,y,'-o')
    plt.xlabel('t')
    plt.ylabel('y')

def f(t, y, r):
    ydot = r * y**2;
    return ydot
```

```python
ode_function()
```

# Logistic Equation

```python
# Logistic equation formula
def f(t, N, r, K):
    Ndot = r * N * (1 - (N / K))
    return Ndot

# Solve it numerically
def ode_function(f, N_0=1, r=0.5, K=10):
    print(f'Parameters used: (N_0, r, K) = ({N_0}, {r}, {K})')
    fig, ax = plt.subplots(1, 2, figsize=(14, 4))
    [t, N_arr] = ode45(lambda t, N: f(t, N, r=r, K=K), [0, 18], N_0)
    ax[0].plot(t, N_arr)
    ax[0].axhline(K, color='grey', alpha=0.5, label='Carrying Capacity')
    ax[0].set(xlabel='t', ylabel='N(t)', title='Numerical Solution', xlim=(0, 17))
    ax[0].legend()
    # Include analytical solution as well
    N_analytical = np.array([K / (1 - ((N_0 - K) / N_0) * np.exp(-r * t_)) for t_ in t])
    ax[1].plot(t, N_analytical, color='crimson')
    ax[1].axhline(K, color='grey', alpha=0.5, label='Carrying Capacity')
    ax[1].set(xlabel='t', ylabel='N(t)', title='Analytical Solution', xlim=(0, 17))
    ax[1].legend()
    
    fig.savefig('numerical_vs_analytical.pdf', bbox_inches='tight')
ode_function(f)
```

We get excellent agreement between the two ways of solving the logistic equation!
