```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
''' Matplotlib customization, from class '''
# Customize for matplotlib
# If interested in the matplotlib object hierarchy, check: https://realpython.com/python-matplotlib-guide/
plt.rcParams['axes.linewidth'] = 1
plt.rcParams['xtick.bottom'] = True
plt.rcParams['ytick.left'] = True
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['mathtext.default'] = 'regular'
# Change font size: http://www.futurile.net/2016/02/27/matplotlib-beautiful-plots-with-style/
plt.rcParams['font.size'] = 12
plt.rcParams['axes.labelsize'] = 14
# plt.rcParams['axes.labelweight'] = 'bold'
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 14
plt.rcParams['figure.titlesize'] = 20
```

```python
# Defining the differential equations of the model
def deriv(t, vals, loads, params):
    # Get values of variables and parameters
    pa, ps, n1, n2 = vals
    # Time-dependent variables
    wa_fun, ws_fun = loads
    wa, ws = wa_fun(t), ws_fun(t)
    # Time-independent parameters
    kappa, tau1, tau2, beta, gamma = params
    
    # Find value of derivatives
    dpa_dt = wa * (1 - kappa * (n2)) - (pa / tau1) - gamma * n1
    dps_dt = ws * (1 - kappa * (n2)) - (ps / tau1) - gamma * n1
    dn1_dt = beta * np.sqrt(t) * (wa + ws) - n1 * (1 + kappa * (pa + ps)) / tau1
    dn2_dt = wa + ws - n2 * (1 + kappa * (pa + ps)) / tau2
    
    return dpa_dt, dps_dt, dn1_dt, dn2_dt
```

```python
# Constant parameters
kappa, tau1, tau2, beta, gamma = 0.02, 21, 1, 0.03, 0.1

# Time-dependent parameters

# Aerobic training
def wa_fun(t):
    if 0 < t % 7 and t % 7 < 1:
        return 0
    return 0.5

# Strength training
def ws_fun(t):
    if 0 < t % 7 and t % 7 < 1:
        return 0
    return 0.5

# Defining arrays of parameters
loads = [wa_fun, ws_fun]
params = [kappa, tau1, tau2, beta, gamma]

# Initial values of (S, I, R) and time information
vals0 = [0, 0, 0, 0]
t = [0, 60]
t_eval = np.linspace(t[0], t[1], 50)

# Solve the system of differential equations
sol = solve_ivp(deriv, t, vals0, args=(loads, params), t_eval=t_eval)
```

```python
# Plot solution
fig, axs = plt.subplots(2, 2,figsize=(8, 6))

# Y-axis labels and titles
ylabs = ["$p_a$", "$p_s$", "$n_1$", "$n_2$"]
# titles = ["Aerobic Fitness", "Strength Fitness", "Long-Term Fatigue", "Short-Term Fatigue"]

counter = 0
for i, ax in enumerate(axs):
    for j in range(2):
        ax[j].plot(sol.t, sol.y[counter], linewidth=1.5, mfc="none", ms=5.5, mew=1.5)
        ax[j].set_ylabel(ylabs[counter], fontweight="bold")
        # ax[j].set_title(titles[counter])
        if i == 1:
            ax[j].set_xlabel("Time [days]", fontweight="bold")
        counter += 1
plt.show()
```
