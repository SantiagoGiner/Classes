# **Task 1:**

We use the analytical solution derived from the notes to model a sprint. 

Below are the analytical solutions to the sprint:

```python
# P: we need to obtain the value of Pmax and plug in here, which is our next task (task 2)
# k:we need to obtain the value of k and plug in here, which is our next task (task 2)
# velocity
v0 = 0 #the initial condition for v is: v(t=0) = 0
def v(t,P,k):
    return P / k * (1 - np.exp(-k * t))

# position
x0 = 0  # the initial condition for x is: x(t=0) = 0
def x(t, P, k):
    return (P / k**2) * (k * t + np.expm1(-k * t))
```

With the velocity and position function, we can get x and v across time. 

# **Task 2:**

---



### Load data

```python
usain_bolt = pd.read_csv("usain_bolt.csv", index_col=None, header="infer")
```

```python
usain_bolt.head()
```

```python
# Shift the time data to account for reaction time
usain_bolt["Time2009_new"] = usain_bolt["Time2009"] - usain_bolt["Time2009"][0]
usain_bolt["Time2008_new"] = usain_bolt["Time2008"]- usain_bolt["Time2008"][0]
```

```python
from scipy.optimize import curve_fit

# Fit curve to data
popt, CovB = curve_fit(x, usain_bolt["Time2009_new"], usain_bolt["Position"])
P_max_fit, k_fit = popt
```

```python
print(f"The parameter Pmax is {P_max_fit},and the parameter k is {k_fit}.")
```

```python
# Plot solution
fig, ax = plt.subplots(1,1,figsize=(9,7))

ax.plot(usain_bolt["Time2009_new"], usain_bolt["Position"], "o", linewidth=1.5, mfc="none", ms=5.5, mew=1.5, label="Data")
ax.plot(usain_bolt["Time2009_new"], x(usain_bolt["Time2009_new"], P_max_fit, k_fit), "--", linewidth=1.5, mfc="none", ms=5.5, mew=1.5, label="Fit")

# Set axes labels and legends
ax.set_xlabel("Time [s]")
ax.set_ylabel("Position [m]")
ax.legend(loc="lower right")

plt.show()
```

As we can see from the the graph, the fitted line and the actual data are quite aligned with each other. 

## **Residual Analysis**
Now we have the parameters Pmax (P) and k, and we will then perform residual analysis. 

```python
residuals = usain_bolt["Position"] - x(usain_bolt["Time2009_new"], P_max_fit, k_fit)
```

First, we plot the residual:

```python
# Plot the residuals
fig, ax = plt.subplots(1, 1, figsize=(9, 7))
ax.plot(usain_bolt["Time2009_new"], residuals, "-o", linewidth=1.5, mfc="none", ms=5.5, mew=1.5)

# Set axes labels and legends
ax.set_xlabel("Time [s]")
ax.set_ylabel("Residual [m]")
plt.show()
```

The residual plot shows that the residuals are all relatively small, fluctuating between -0.15 and 0.25. 

```python
# Compute the mean squared error
MSE = residuals.T @ residuals / (usain_bolt.shape[0] - 2)
print("The mean squared error is:", MSE)

# Print covariance matrix
print("The covariance matrix for the parameters", CovB)
```

The mean squared error is also relatively small, and the value in the covariance matrix is also quite small. 

Next, check autocorrelation

```python
# https://www.statsmodels.org/dev/generated/statsmodels.graphics.tsaplots.plot_acf.html
from statsmodels.graphics.tsaplots import plot_acf

# Plot the autocorrelation
fig, ax = plt.subplots(1, 1, figsize=(9, 7))
fig = plot_acf(residuals, lags=10, ax=ax)

# Set title, axes labels and legends
ax.set_title("Sample autocorrelation function")
ax.set_xlabel("Lag")
ax.set_ylabel("Sample Autocorrelation")
plt.show()
```

The autocorrelation plot indicates that autocorrelation mitigates across time. Tus the uncertainty estimates mitigates across time. 

### Confidence intervals (95% confidence intervals): Estimation of Parameter Uncertainties

```python
# http://scipy.github.io/devdocs/reference/generated/scipy.stats.t.html
from scipy.stats.distributions import t

alpha = 0.05
dof = usain_bolt.shape[0] - 2 # statistical degree of freedom
tval = t.ppf(1.0 - alpha / 2., dof)
```

```python
ci = np.zeros((2,2))
for i in range(2):
    ci[i,0] = popt[i] - CovB[i,i]**0.5 * tval
    ci[i,1] = popt[i] + CovB[i,i]**0.5 * tval
```

The 95% confidence interval for Pmax is (9.6652176 , 10.03123839), and the 95% confidence interval for k is (0.78698189,0.82443147). Both confidence intervals are pretty narrow, suggesting that the estimation of parameters are relatively certain.

```python
from scipy.optimize import fsolve

race_distances = np.array([200, 400, 800, 400, 1000, 1500, 2000, 3000, 5000, 10000, 21097, 42194])
pred_times = np.zeros(len(race_distances))
for i, race_dist in enumerate(race_distances):
    solve = lambda t: x(t, popt[0], popt[1]) - race_dist
    pred_times[i] = fsolve(solve, 20)
```

```python
for time, dist in zip(pred_times, race_distances):
    print(f"Predicted time for {dist} meters: {time} s = {time / 60} min") 
```

The computed results above is faster than the world record. This make sense as 
after some critical distance, the athlete’s initial reserve energy E0 (per unit mass) will be exhausted and continued energy expenditure will need to be supplied through the oxygen uptake, which is assumed to be constant at the maximum oxygen uptake again on the ground that the athlete is making their maximum effort. Hence, the longer than running distance, the more inaccurate our model will be as athlete is getting more and more exhausted.
