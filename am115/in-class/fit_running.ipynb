## Part 2

Let us first load the data from the provided csv file

```python
# Load data from usain_bolt.csv
times, positions = [], []
with open("usain_bolt.csv") as f:
    reader = csv.reader(f)
    # Ignore row with column names
    next(reader)
    for row in reader:
        times.append(float(row[1]))
        positions.append(float(row[0]))
times, posistions = np.array(times), np.array(positions)
```

```python
""" Adapted from fit_yeast_ode.ipynb """
# Define exact solution
exact_y = lambda t, P_max, k: (P_max / k**2) * (k *  t + np.expm1(-k * t))

# Define curve to be fitted
def exact_solution(times, P_max,  k):
    y_fit = interp1d(times, np.array([exact_y(t, P_max, k) for t in times]))
    return y_fit(times)

# Define the residuals -> difference between prediced and exact
def residuals(coeffs, y, t):    
    return y - exact_solution(t, coeffs[0], coeffs[1])

# Fir a nonlinear model
initial_guess = [1, 1]
nlinfit = least_squares(residuals, initial_guess, args=(positions, times), method='lm')
beta = nlinfit.x
```

Print the coefficients found to best fit the running data

```python
# Print the coefficients found to be the best
print(f"The best P_max was found to be: {beta[0]}")
print(f"The best k was found to be: {beta[1]}")
```

Now let us calculate the predicted positions for the array of times

```python
best_fit = exact_solution(times, beta[0], beta[1])

# Plot solution
fig, ax = plt.subplots(1,1,figsize=(10, 8))
ax.plot(times, positions, 'o', linewidth=1.5, mfc="none", ms=5.5, mew=1.5, label="Data")
ax.plot(times, best_fit, '--', linewidth=1.5, mfc="none", ms=5.5, mew=1.5, label="Fitted Curve")

# Set axes labels and legends
ax.set_xlabel("Time")
ax.set_ylabel("Position")
ax.legend()
plt.show()
```

## Part 3
Let us compute the predicted running time to complete 200 meters, 400 meters, 800 meters, etc., up to the marathon, 42,200 meters.

```python
from scipy.optimize import fsolve

race_distances = np.array([200, 400, 800, 400, 1000, 1500, 2000, 3000, 5000, 10000, 21097, 42194])
pred_times = np.zeros(len(race_distances))
for i, race_dist in enumerate(race_distances):
    solve = lambda t: exact_y(t, beta[0], beta[1]) - race_dist
    pred_times[i] = fsolve(solve, 20)
```

```python
for time, dist in zip(pred_times, race_distances):
    print(f"Predicted time for {dist} meters: {time} s = {time / 60} min") 
```

The predicted times are clearly much lower than the actual values, due to the fact that in our model, the $P_\text{max}$ acceleration is constant, which would clearly not be the case in the real world, since the acceleration of a sprinter is much higher than that of a marathon runner, as the marathon runner must provide a consistent acceleration through a much larger distance and would therefore tire very quickly if they were to try to sustain the acceleration of a 100-m sprinter over a longer race.
