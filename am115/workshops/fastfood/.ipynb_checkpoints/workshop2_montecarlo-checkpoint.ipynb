This situation can be modeled through a Poisson process. Let us define $\lambda$ as the average rate at which a customer arrives. Then, in the Poisson process model, we assume that the number of customers that arrive in some time $t$ is distributed as Pois($\lambda t$). In this scenario we would consider the number of arrivals in disjoint time intervals as independent. This tells us that the times between customer arrivals, i.e. the inter-arrival times, are distributed as i.i.d. Expo($\lambda$).<br><br>

For service time, we can assume the service time is Expo($\lambda$) as well.  However, given more information, other distributions could be used.  For example, we could average the time it takes to make each meal on the menu over several trials and model the service time as a Normal distribution with mean given by this sample mean and variance given by the sample variance of the meals' cooking times.

a. Assuming that the service time is exponentially distributed does not seem to be a very good approximation of what the real world service time would be. We can see this by noting that the memoryless property, a key feature of exponential distributions, is highly unlikely to hold in this real-life fast food scenario.  Memorylessness implies that if a customer has waited a certain length of time, the expected additional waiting time is still the theoretical mean of the distributios, but this doesn't make sense because in reality, for if you have waited a long time for your order, you would expect that your order should come soon.  In a Poisson process, we also assume the number of arrivals in disjoint time intervals to be independent.  This often does not hold because people are less likely to come into empty restuarants or people will arrive with their friends, creating dependence between events.

The expected total wait time (wait time plus service time) is $\frac{1}{b-a}$.  For $a = 1$ and $b = 1.5$ this means the expected total wait time is $2$. 

To see this, the probability that total time $T$ exceeds a value $t$ is given by

$$ P(T>t) = \exp(-(b-a)t). \tag{1}$$

From this, we can compute the expected total time through integration by parts. 
 
From Eq.(1), the probability of $T$ being in the range $[t,t+\text{d}t]$ is: 

$$ P(t<T<t+\text{d}t) = (b-a)\exp(-(b-a)t)\text{d}t. $$

Then the expected time is

$$ \int_{0}^{\infty} t P(t<T<t+\text{d}t) = \int_0^{\infty} (b-a)t\exp(-(b-a)t)\text{d}t. $$

Define $x:=(b-a)t$, we have

$$ \int_{0}^{\infty} t P(t<T<t+\text{d}t) = \frac{1}{b-a}\int_0^{\infty}x\exp(-x)\text{d}x = \frac{1}{b-a}, $$

where in the last step we used the fact that $x \exp(-x)$ is the PDF of the Expo$(1)$ distribution, and the integral of a PDF over its support is $1,$ by definition of the PDF.

The following code verifies our result. 

```python
a = 1 # average number of arrivals per minute 
b = 1.5 # average number of people served per minute (reduce from 1.5 to 1.1 or 0.95)
ncust = 10000 # number of customers 
```

```python
# Initialize arrays 
at = np.zeros((ncust, 1))
ft = np.zeros((ncust, 1))
# Generate random arrival times assuming Poisson process:
r = np.random.uniform(low=0., high=1., size=(ncust, 1))
iat = -(1./a) * np.log(r) # generate inter-arrival times, exponential distribution
at[0] = iat[0] # arrival time of first customer
for i in range(1, ncust): 
    at[i] = at[i-1] + iat[i]
    
# Generate random service times for each customer:
r = np.random.uniform(low=0., high=1., size=(ncust, 1))
st =  -(1./b) * np.log(r) # service time for each customer, follow an expoential distribution
# Compute time each customer finishes:
ft[0] = at[0] + st[0]
for i in range(1, ncust): 
    # compute finish time for each customer as the larger of
    # arrival time plus service time (if no wait)
    # finish time of previous customer plus service time (if wait)
    ft[i] = np.maximum(at[i] + st[i], ft[i-1] + st[i])
    
total_time = ft - at # total time spent by each customer
wait_time = total_time - st # time spent waiting before being served
ave_service_time = np.sum(st)/np.float32(ncust)
ave_wait_time = np.sum(wait_time)/np.float32(ncust)
ave_total_time = np.sum(total_time)/np.float32(ncust)
print('ave_service_time = ', ave_service_time)
print('ave_wait_time = ', ave_wait_time)
print('ave_total_time = ', ave_total_time)
```

From the output we can see the estimation of expected total wait time is about $2$, which is consistent with our analysis. 

b. We choose the guaranteed time by considering its relationship with inter-arrival time.  With a low guaranteed time, customers can assume they are more likely to get free food, as it is more likely that the restaurant cannot serve them in very short times.  This will drive more customers to the store, and therefore decrease inter-arrival time.  By the same principle, for higher guaranteed times, customers will assume they are less likely to get a free meal, meaning that less customers will come to the store and inter-arrival times will increase.  This increase will continue until some fixed point, where the possibly of getting a free meal hardly drives any more customers to the store, and customers are simply coming because they enjoy the food (and are willing to pay full price with near certainty).  As such, the inter-arrival time will begin to plateau at a value around what the baseline rate would be with no guarantee time whatsoever.  One example of this could a sigmoid function, which we consider below.

```python
# Array of different time guarantees
u_arr = np.linspace(-5, 30, 100)

#relating guaranteed time to intervarrival time through a sigmoid
k = 0.35
a = 10.5
i_at = 10 * 1 / (1 + np.exp(-k * (u_arr - a)))

# Plotting the curve
plt.figure(figsize = (8,4))
plt.plot(u_arr, i_at)
plt.xlim(0,27)
plt.ylim(0,11)
plt.title('Interarrival Time against Guaranteed Time Modeled by Sigmoid')
plt.xlabel('guaranteed time (mins)')
plt.ylabel('interarrival time (mins)')
plt.show()
```

As the plot shows, with low guarantee times, inter-arrival times will also be low (but never completely reaching $0$ as this is unrealistic).   Around $20$ minutes of guaranteed time, we see a plateau in interarrival time begin to occur, as a high guaranteed time is unlikely to drive significantly many more customers to the door than if there were no guarantee.  While the curve gets very flat, it never plateaus completely, representing how, even if very large arrival times, there may still be a few more customers that want to come to the store (lowering interarrival time ever so slightly) than if there were no guarantee time whatsoever.  (Note: the times chosen for this example are arbitrary and were simply used to demostrate how a sigmoid curve relatig guaranteed time to interarrival time could arise)<br><br>



We can also take a more simplified approach where we model the mean of the inter-arrival times as a piecewise-linear function of the time guarantee.  Here, the inter-arrival time, $\frac{1}{a}$, behaves as:

$$
\frac{1}{a(u)} = \begin{cases}
\frac{\gamma u}{u_0} \text{ if } \frac{u_0}{b\gamma} < u < u_0 \\
\gamma \text{ if } u_0 < u < \infty
\end{cases}
$$

This is essentially a more crude representation of the same trends we see above with the sigmoid.  It indicates that the inter-arrival time increases linearly as the time  guarantee increases up to the point of $u_0$ and that time guarantees longer than $u_0$ do not have an impact on customer behavior. The first inequality ensures $\frac{1}{a} > \frac{1}{b}$ such that the queue cannot become infinitely long. (note: this simplification is pulled from class notes)

```python
#defining the piecewise linear function
def piecewise_intarr(ts,u0):
  res = []
  for t in ts:
   res.append(0.52*t) if t < u0 else res.append(res[-1])
  return res

# Plotting the curve
plt.figure(figsize = (8,4))
plt.plot(u_arr, piecewise_intarr(u_arr,19))
plt.xlim(0,27)
plt.ylim(0,11)
plt.title('Interarrival Time against Guaranteed Time Modeled by Piecewise Linear Function')
plt.xlabel('guaranteed time (mins)')
plt.ylabel('interarrival time (mins)')
plt.show()
 
```

c.Assuming we know the three following quantities:

The industry average arrival time: $a$.  **Note: $a$ is not a constant but a function of the guaranteed time $u$ so $a=a(u)$**

The industry average service time $b$.

We could use the following strategy to get the expected profit given the guaranteed time:<br><br>

First, let $p$ be the price divided by cost and call $(p-1)$ the profit margin.  If we let $u$ be the time guarantee, then we have the profit divided by the cost:

$$
P(T) = \begin{cases}
p-1 \text{ if } T \leq u, \\
-1 \text{ otherwise.}
\end{cases}
$$

From here we can derive the expected profit given the guaranteed time ($u$) as:

$$
\begin{split}
E[P] &= a[(p-1) P(T \leq u) - P(T \geq u)] \\
&= a[(p-1)(1-\exp(-(b-a)u))-\exp(-(b-a)u)] \\
&= a[p - 1 - p \cdot \exp(-(b-a)u)]
\end{split}
$$

so the expected profit (divided by the cost) per unit time as a function of the guaranteed time $(u)$ is:
$$J(u) = a(u)[p - 1 - p \exp(-(b-a(u))u)]$$

In the case where the interarrival time behaves piecewise linearly (following the equation shown in part b), we have that

$$J(u) = \frac{u_0}{\gamma u}\left[p - 1 - p \exp\left(-bu + \frac{u_0}{\gamma}\right)\right] \text{ if } \frac{u_0}{b\gamma} < u < u_0,$$

and

$$ J(u) = \frac{1}{\gamma} \left[p - 1 - p \exp\left(-bu + \frac{u}{\gamma}\right)\right] \text{ if } u_0 < u < \infty. $$

```python
# Customize for matplotlib
# If interested in the matplotlib object hierarchy, check: https://realpython.com/python-matplotlib-guide/
plt.rcParams['axes.linewidth'] = 1
plt.rcParams['xtick.bottom'] = True
plt.rcParams['ytick.left'] = True
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['mathtext.default'] = 'regular'
# Change font size: http://www.futurile.net/2016/02/27/matplotlib-beautiful-plots-with-style/
plt.rcParams['font.size'] = 12
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['axes.labelweight'] = 'bold'
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 14
plt.rcParams['figure.titlesize'] = 20
```

```python
# Industy standard 
a = 1  # average number of arrivals per minute 
b = 1.5  # average number of people served per minute (reduce from 1.5 to 1.1 or 0.95)
ncust = 10000  # number of customers 
```

```python
# Set the parameters
u0 = 1
gamma = 1/a
u = np.linspace(u0/(b*gamma),20,1000)
n = len(u)

# Arrays to store the profits
profit_list = np.zeros((n,1))
profit_truth = np.zeros((n,1))
p = 2

# Mean of inter-arrival times
def iat(u):
    if u >= u0:
        return gamma
    elif u0 / (b * gamma) <= u < u0:
        return gamma * u/u0

# Profit function
def profit(T,u):
    if T <= u:
        return p - 1
    else:
        return -1

# Iterate over values of u
for j in range(n):
    # Initialize arrays 
    at = np.zeros((ncust, 1))
    ft = np.zeros((ncust, 1))

    # Generate random arrival times assuming Poisson process:
    r = np.random.uniform(low=0., high=1., size=(ncust, 1))
    iat_ = -iat(u[j]) * np.log(r)  # generate inter-arrival times, exponential distribution
    at[0] = iat_[0]  # arrival time of first customer
    for i in range(1, ncust): 
        at[i] = at[i-1] + iat_[i]

    # Generate random service times for each customer: 
    r = np.random.uniform(low=0., high=1., size=(ncust, 1))
    st = -(1./b) * np.log(r)  # service time for each customer, follow an expoential distribution

    # Compute time each customer finishes:
    ft[0] = at[0] + st[0]
    profit_ = 0
    for i in range(1, ncust): 
        # compute finish time for each customer as the larger of
        # arrival time plus service time (if no wait)
        # finish time of previous customer plus service time (if wait)
        ft[i] = np.maximum(at[i] + st[i], ft[i-1] + st[i])
    for i in range(ncust):
        profit_ += profit(ft[i] - at[i], u[j])
    profit_list[j] = profit_ / ft[-1]
    ar = 1 / iat(u[j])
    profit_truth_ = ar * (p - 1 - p * np.exp(-(b - ar) * u[j]))
    profit_truth[j] = profit_truth_
```

```python
# Plot profit vs time guarantee
fig, axs = plt.subplots(1,1,figsize=(9,7), tight_layout=True)
axs.plot(u, profit_list, color='b', label='simluation')
axs.plot(u, profit_truth, color='r', label='analytical solution')
axs.set(ylabel = 'profit (per minute)')
axs.set(xlabel = 'time guarantee')
axs.set_title("profit vs. time guarantee")

plt.legend()
plt.show()
```

In this plot, we can see when the time guarantee $u$ increases, the profit also increases. Here we set time guarantee in the interval $\left[\frac{u_0}{b\gamma},20\right]$, and profit is $1$ if the service time is in the range of guarantee, otherwise the profit is $-1$. And for each value of $u$, we run $10000$ trials, in each trial, we assume there is $10000$ customer arrive. And the simulation result is based on the average of all the trials. Therefore, it is valid and the profit is close to the truth value. Since the profit curve shows an increasing trend and then flattens after 10, the value of $u$ to maximize the profit is $>=10$ mins, which means if we want the maximum profit, we should set the guarantee time to be approximately greater than 10 mins **for the current industry standard inter-arrival and service time**. However, given that the general plot of this graph is that the longer the time guarantee time the more profits fast food service can make, the optimal strategy is still do not set a guarantee time.

## Sigmoid Model for Inter-Arrival Times

```python
# Industy standard 
a = 1  # average number of arrivals per minute 
b = 1.5  # average number of people served per minute (reduce from 1.5 to 1.1 or 0.95)
ncust = 10000  # number of customers 
```

```python
# Set the parameters
u0 = 1
gamma = 1 / a
u = np.linspace(u0 / (b * gamma), 20, 1000)
n = len(u)

# Arrays to store the profits
profit_list = np.zeros((n,1))
profit_truth = np.zeros((n,1))
p = 2

# Mean of inter-arrival times as a sigmoid function
k = 0.35
iat = lambda t: 10 * 1 / (1 + np.exp(-k * (t - a)))

# Profit function
def profit(T, u):
    if T <= u:
        return p - 1
    else:
        return -1

# Iterate over values of u
for j in range(n):
    # Initialize arrays 
    at = np.zeros((ncust, 1))
    ft = np.zeros((ncust, 1))

    # Generate random arrival times assuming Poisson process:
    r = np.random.uniform(low=0., high=1., size=(ncust, 1))
    iat_ = -iat(u[j]) * np.log(r)  # generate inter-arrival times, exponential distribution
    at[0] = iat_[0]  # arrival time of first customer
    for i in range(1, ncust): 
        at[i] = at[i-1] + iat_[i]

    # Generate random service times for each customer: 
    r = np.random.uniform(low=0., high=1., size=(ncust, 1))
    st = -(1. / b) * np.log(r)  # service time for each customer, follow an expoential distribution

    # Compute time each customer finishes:
    ft[0] = at[0] + st[0]
    profit_ = 0
    for i in range(1, ncust): 
        # compute finish time for each customer as the larger of
        # arrival time plus service time (if no wait)
        # finish time of previous customer plus service time (if wait)
        ft[i] = np.maximum(at[i] + st[i], ft[i-1] + st[i])
    for i in range(ncust):
        profit_ += profit(ft[i] - at[i], u[j])
    profit_list[j] = profit_ / ft[-1]
    ar = 1 / iat(u[j])
    profit_truth_ = ar * (p - 1 - p * np.exp(-(b - ar) * u[j]))
    profit_truth[j] = profit_truth_
```

```python
# Plot profit vs time guarantee
fig, axs = plt.subplots(1,1,figsize=(9,7), tight_layout=True)
axs.plot(u, profit_list, color='b', label='simluation')
axs.plot(u, profit_truth, color='r', label='analytical solution')
axs.set(ylabel = 'profit (per minute)')
axs.set(xlabel = 'time guarantee')
axs.set_title("profit vs. time guarantee")

plt.legend()
plt.show()
```

This model reaches a maximum profit value for a specific value of the time guarantee, which occurs at around $u = 2.5,$ for the current values of $a = 1$ and $b = 1.5.$

Note that this sigmoid model assumes that the mean of the inter-arrival times is not dependent on $b,$ the average number of customers served per unit time.
