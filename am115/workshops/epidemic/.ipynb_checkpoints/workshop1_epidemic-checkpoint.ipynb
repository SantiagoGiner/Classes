```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

## Problem 1

```python
#defining the SIR system of equations
def deriv(t, z, a, b):
    S, I, R = z
    N = S + I + R
    
    dS = -a/N * I * S
    dI = a/N * I * S - b*I
    dR = b * I
    
    return dS, dI, dR 

#defining values for parameters
a = 2 * np.log(2)
b = np.log(2)

#defining initial populations
S = 999
I = 1
R = 0

#defining initial conditions and time interval
y0 = (S,I,R)
t = [0, 20]
t_eval = np.linspace(0, 20, 100)

#solving system of differential equations
sol = solve_ivp(deriv,t,y0,args = (a,b), t_eval=t_eval)
```

```python
# Plot solution
fig, ax = plt.subplots(1,1,figsize=(9,7))

ax.plot(sol.t, sol.y[0], '-o', linewidth=1.5, mfc='none', ms=5.5, mew=1.5,  label='Susceptible')
ax.plot(sol.t, sol.y[1], '-s', linewidth=1.5, mfc='none', ms=5.5, mew=1.5,  label='Infected')
ax.plot(sol.t, sol.y[2], '-^', linewidth=1.5, mfc='none', ms=5.5, mew=1.5,  label='Recovered')

# Set axes labels and legends
ax.set_xlabel('Time')
ax.set_ylabel('Population')
ax.set_title("SIR Populations from $t = 0$ to $t = 20$")
ax.legend(loc='center left')

plt.show()
```

We expect the initial growth of the infected population to be exponential since initially the number of infected $I$ is negligble compared to the number of susceptible, meaning that we initially have that the rate of change of infected is approximately $$\frac{dI}{dt} \approx \frac{a}{N} IS,$$ which corresponds to exponential growth of $I.$

But this exponential growth will not continue indefinitely, since as $I$ increases, the rate of change of $S$ becomes more negative, i.e. $S$ decreases more rapidly, since $dS/dt \propto -I$. There will then be a point where $S$ the $-bI$ term will dominate in the equation for $dI/dt,$ corresponding to expinential decrease, and so $I$ will start to decrease. Solving for $dI/dt = 0$ will give us the number of susceptible population that corresponds to the rate of change of $I$ going from positive to negative, which means that $I$ would start to decrease after this point. Solving this equation, we get $$S = \frac{b}{a}N.$$

We can verify this by calculating what this value corresponds to in our simulation.

```python
# Value of S after which I starts to decrease
N = 1000
b * N / a
```

This makes sense: if we look at the point in the plot above where $S \approx 500,$ this also corresponds to the point in time when $I$ starts to decrease.

```python
fig, ax = plt.subplots(1,4,figsize=(18,6))
fig.suptitle('SIR Model with varying values of R0',fontsize = 16)

multiplier = [2,2.5,3,5]

#looping through different values for a
for i, m in enumerate(multiplier):
    a = m * np.log(2)
    
    #solving systems with differnet a values
    sol = solve_ivp(deriv,t,y0,args = (a,b), t_eval=t_eval)

    # Plot solution
    ax[i].plot(sol.t, sol.y[0], '.-', linewidth=1.5, mfc='none', ms=5.5, mew=1.5,  label='Susceptible')
    ax[i].plot(sol.t, sol.y[1], '.-', linewidth=1.5, mfc='none', ms=5.5, mew=1.5,  label='Infected')
    ax[i].plot(sol.t, sol.y[2], '.-', linewidth=1.5, mfc='none', ms=5.5, mew=1.5,  label='Recovered')

    # Set axes labels and legends
    ax[i].set_title(f'$R_0$ = {a/b}')
    ax[i].set_xlabel('Time')
    ax[i].set_ylabel('Population')
    ax[i].legend(loc='best')
    
plt.show()
```

As $R_0$, the number of people an infected person can infect before they recover, increases the evolution of the system changes.  For smaller values of $R_0$, the infected population does not spike intially and there is a more gradual decrease in the susceptible population.  Because $R_0$ is small, eventually system reaches a point of relative stability, where a portion of the susceptible population has never been infected, but this number is small enough, and interaction between infected and susceptible individuals is low enough that there is rarely ever any more transmission of the disease.  On the other hand, when $R_0$ is high, we observe an immediate spike in the infected population (at the same time as a harsh drop in the susceptible population), because each infected individual has the capacity to infect many others.  Unlike with smaller $R_0$ values, in this situation, every susceptible individual will end up getting the disease within a very quick time frame.  In reality this could lead to overwhelming hospitals and medical facilities.  Looking at these analyses provides support for measures like social distancing, where we aim to keep smaller $R_0$ values by limiting the number of people an infected person can interact with and infect.

## Problem 2

We consider a time interval h small enough where only 1 infection or 1 recovery (or neither) can occur.  We then look to find the following probabilities:

The probability an infection occurs in the interval h:
$P(S(t+h) = S-1 \cap R(t) = R | S(t) = S \cap R(t) = R)$

The probability a recovery occurs in the interval h:
$P(S(t+h) = S \cap R(t+h) = R+1| S(t) = S \cap R(t) = R )$  

The probability neither occurs in the interval h:
$P(S(t+h) = S \cap R(t+h) = R | S(t) = S \cap R(t) = R)$

<br>

Let $p_s$ be the probability that a suspceptible person gets infected over a sufficiently small time interval h.

By law of large numbers (LLN), with a large enough susceptible population, $S$, the number of  individuals that get infected over a time interval h, with  certainty, is $p_shS$.

Thus, the rate at which $S$ decreases can be desribed as 
$- \frac{d}{dh} (p_shS) = -p_sS $


Setting this equal to $\frac{dS}{dt} = -\frac{a}{N}IS$ gives  

Setting this equal to $-p_sS = -\frac{a}{N}IS$ so

$$p_s = \frac{a}{N}I$$

Therefore, assuming independence between infections, the probability of the susceptible population of size S decreasing by 1 over a sufficiently small time interval $h$ will be: $$p_sS = \frac{a}{N}IS$$. <br><br><br>



Similarly, let $p_r$ be the probability that an individual recovers from infection in a sufficiently small time interval $h$.

Again, by the LLN, the number of recovered individuals $R$ increases by $p_rhR$ for sufficiently large $R,$ over a time interval $h.$ Therefore, the rate of change of the recovered population in this probabilistic model of the recovered population is $p_rI,$ since every infected individual $I$ can recover with probability $p_r$ in the time $h.$

To be consistent with the deterministic approach, we demand that this rate of change be equal to the time derivative of $R$ in the SIR model, that is,

$$p_rI = bI \implies p_r = b.$$

Thus, assuming indepence between recoveries, the probability that the recovered population $R$ increases by sufficiently small time interval $h$ is: $$p_rI = bI$$.<br><br><br>

Hence, it follows that the probability that nothing happens, i.e., that all populations stay the same over some sufficiently small time interval is,

$$1 - (p_sS + p_rI)$$

with $p_s$ and $p_r$ defined as above.<br><br><br>


With these values, we can simulate the evolution of the populations probabilistically.

```python
""" Code adapted from notebook provided for HW 2 """
# Set initial population and steps for simulation
N = 1000
n_steps = 23000
dt = 10**(-3)

# Initialize arrays for S, I, R
S, I, R = np.zeros(n_steps), np.zeros(n_steps), np.zeros(n_steps)
S[0] = 999
I[0] = 1

# Constants of the system's evolution
a = 2 * np.log(2)
b = np.log(2)

# Iterate until n_steps is reached
t = 0
while t < n_steps - 1 and I[t] >= 0:
    ps = a * I[t] / N
    pr = b
    # Generate random probability
    r = np.random.rand(1)
    # Infection occurs
    if r < ps * S[t] * dt:
        S[t + 1] = S[t] - 1
        I[t + 1] = I[t] + 1
        R[t + 1] = R[t]
    # Recovery occurs
    elif r < (ps * S[t] + pr * I[t]) * dt:
        S[t + 1] = S[t]
        I[t + 1] = I[t] - 1
        R[t + 1] = R[t] + 1
    # Else, nothing happens
    else:
        S[t + 1] = S[t]
        I[t + 1] = I[t]
        R[t + 1] = R[t]
    t += 1

# Plot solution
times = np.linspace(0, n_steps, n_steps)
fig, ax = plt.subplots(1, 1, figsize=(9, 7))
ax.plot(times, S,'.-', linewidth=1.5, mfc='none', ms=5.5, mew=1.5,  label='Susceptible')
ax.plot(times, I,'.-', linewidth=1.5, mfc='none', ms=5.5, mew=1.5,  label='Infected')
ax.plot(times, R,'.-', linewidth=1.5, mfc='none', ms=5.5, mew=1.5,  label='Recovered')

# Set axes labels and legends
ax.set_xlabel('Time Steps')
ax.set_ylabel('Population')
ax.set_title(f"Stochastic SIR Populations from $t = 0$ to $t = {n_steps}$")
ax.legend(loc='center left')

plt.show()
```

Looking side-by-side at the two epidemic evolution models, its clear the results show consistency.  In both the deterministic and stochastic approaches we observe a gradual, then increased, then gradual again, decrease in the susceptible population; an initial rise and then fall in the infected population; and a gradual, then increased, then gradual again, increase in the recovered population.  The curves follow nearly identical trends despite our use of different modeling techniques. <br>

Thse evolutions coincide with our intution about the problem.  No matter the approach, we would expect the growth of the infected population to increase slowly initially (as there are few infected individuals), then rise more quickly as a higher number of individuals become infected (but there are still many susceptible individuals left), and the fall again when the population of susceptible (never-been-infected) people reaches a low enough point that it is hard to find people to infect.  The population curves for susceptible and recovered follow paths that complement this activity in the evolution of infections.





## Problem 4

```python
#the following code is adapted from the class code "model_epidemic_1d.ipynb" - individual agent model 
def simulate1D(n, nc, pt):
    """Package the simulation steps of intializng the population
    vector and iterating the epidemic and census functions unit I=0
    the censuses are accumulated in a results matrix

    Args:
        n (int): time
        nc (numpy.ndarray): population
        pt (float): parameter value

    Returns:
        tt (list): time
        results (list): sir subpopulation at each time
        
    """
    
    t = 0
    # Set up an intial population of susceptibles
    pop = initial1D(n,0,0)
    # Put an infected individual in the middle
    pop[int(n/2)-1] = 'i'
    s,i,r = census1D(pop)
    # Start the results mnatrix
    results = [[s,i,r]]
    tt=[t]
    
    while i>0:
        pop = epidemic1D(nc,pt,pop)
        s,i,r = census1D(pop)
        results.append([s,i,r])
        t = t+1
        tt.append(t)
        
    return tt,results

def initial1D(s0,i0,r0):
    """Set an initial population vector for the epidemic simulation;
    each state subpopulation is appended to the growing vector, pop
    
    NOTE: if position in the array is important,eg. if you are modeling
    local neighborhood contacts, each individual must be placed 
    randomly in the array

    Args:
        s0 (int): initial population of susceptibles
        i0 (int): initial population of infectious
        r0 (int): initial population of recovered

    Returns:
        list: net population vector of subpopulation of each state
        
    """
    
    pop = []
    for i in range(s0):
        pop.append('s')
    for i in range(i0):
        pop.append('i')
    for i in range(r0):
        pop.append('r')
        
    return pop
def census1D(pop):
    """Count the number of s, i and r cells in pop

    Args:
        pop (list): net population vector of subpopulation of each state

    Returns:
        s (int): number of susceptibles
        i (int): number of infectious
        r (int): number of recovered
        
    """
    
    s = 0
    i = 0
    r = 0
    # pop is a 1 by n array 
    n = np.array(pop).shape[0]
    
    for j in range(n):
        if pop[j]=='s':
            s += 1
        elif pop[j]=='i':
            i += 1
        elif pop[j]=='r':
            r += 1
            
    return s,i,r
def epidemic1D(nc, pt, pop1):
  """Calculate a new pop2 vector of 's', 'i', and 'r' from 
  the current pop1 assuming that all 'i' cells go to 'r' 
  after each has nc random infectious contacts with other 
  cells. An 's' cell goes to 'i' with probability, pt, if 
  it is contacted by an 'i'

  Args:
      nc (int): all 'i' cells go to 'r' after random 
                infectious contacts with other cells
      pt (float): probability of 's' cell going to 'i'
      pop1 (list): current net population vector

  Returns:
      list: new net population vector
      
  """
  
  pop2 = pop1.copy() # Set up new vector
  n = np.array(pop1).shape[0]
  for j in range(n):
      if pop1[j]=='i':
          for c in range(nc): # nc contact
              k = j
              while k==j: # try to generate a random k NOT equal to j
                  k = np.random.randint(n) # random individual to contact
              if pop1[k]=='s' and np.random.rand()<pt: # rand<pt generates a Bernoulli trial
                  pop2[k] = 'i' # infect the 's' cell
          pop2[j] = 'r' # then the jth individual recovers  
          
  return pop2
```

```python
tt,results=simulate1D(9999,4,0.5)
results=np.array(results) # change results to numpy.ndarray for plotting purposes
```

```python
# Plot solution
fig, ax = plt.subplots(1,1,figsize=(9,7))

ax.plot(tt, results[:,0], '-o', linewidth=1.5, mfc='none', ms=5.5, mew=1.5,  label='susceptible')
ax.plot(tt, results[:,1], '-s', linewidth=1.5, mfc='none', ms=5.5, mew=1.5,  label='infectious')
ax.plot(tt, results[:,2], '-^', linewidth=1.5, mfc='none', ms=5.5, mew=1.5,  label='recovered')

# Set axes labels and legends
ax.set_xlabel('time')
ax.set_ylabel('population')
ax.legend(loc='center left')

plt.show()
```

## Problem 5

A Deterministic Model allows you to calculate a future event (the susceptible/infectious/recovered population) deterministically without the involvement of randomness. We have all the data necessary to predict (determine) the outcome with certainty.

A Stochastic Model can handle uncertainties in the inputs ((infectious rate/recovered rate) applied. Stochastic models possess some inherent randomness - the same set of parameter values and initial conditions will lead to an ensemble of different outputs.

Deterministic models have the benefit of simplicity. They are easier to understand for some of the audience. However, the deterministic model relies on an accurate assessment of the infectious/recovered rate and needs to construct based on rigid assumptions. A Stochastic Model has more tolerance over the uncertainty here but is more complicated and difficult to understand.  In both cases we are forced to make simplifying assumptions that reduce the complexity of our model, but impact their direct applicability.  For example, the in probabilistic approach, we assume independence between infection events.  In reality, this may not be the case (ie. often family members tend to get exposed and fall ill at the same time).

The individual agent model is a simulation model that could require large computational power and computational time if the scenarios are complicated. However, the advantage of the individual agent model comes in its adaptability. We can include different additional parameters without creating a new equation-based model. Using such a simulation model allows one to capture the stochasticity in a real-world system.
