```python
# Load data into NumPy arrays
# Raw amplitude features
Xtrain_amp = np.load("Xtrain_amp.npy")
ytrain_amp = np.load("ytrain_amp.npy")
Xtest_amp = np.load("Xtest_amp.npy")
ytest_amp = np.load("ytest_amp.npy")

# Mel spectrogram features
Xtrain_mel = np.load("Xtrain_mel.npy")
ytrain_mel = np.load("ytrain_mel.npy")
Xtest_mel = np.load("Xtest_mel.npy")
ytest_mel = np.load("ytest_mel.npy")

# Flatten the Mel spectrogram features
Xtrain_mel = Xtrain_mel.reshape(Xtrain_mel.shape[0], -1)
ytrain_mel = ytrain_mel.reshape(ytrain_mel.shape[0], -1)
Xtest_mel = Xtest_mel.reshape(Xtest_mel.shape[0], -1)
ytest_mel = ytest_mel.reshape(ytest_mel.shape[0], -1)
```

```python
# Shape comparisons
print(f"X_train amp shape: {Xtrain_amp.shape}")
print(f"X_train mel shape: {Xtrain_mel.shape}")
```

```python
# Class names and frequencies

# Names of every class
class_names = [
    "air_conditioner",
    "car_horn",
    "children_playing",
    "dog_bark",
    "drilling",
    "engine_idling",
    "gun_shot",
    "jackhammer",
    "siren",
    "street_music"
]

# Corresponding frequencies
class_freqs = [
    12.6,
    3.54,
    12.53,
    9.42,
    10.93,
    12.98,
    1.49,
    11.85,
    12.03,
    12.61
]

# Helper functions

# Function to ouput the accuracy of a model
def print_accuracies(y_train_pred, y_train, y_test_pred, y_test):
    print("Mean accuracy on training data: ", round(accuracy_score(y_train, y_train_pred), 3))
    print("Balanced accuracy on training data: ", round(balanced_accuracy_score(y_train, y_train_pred), 3))
    print("Mean accuracy on test data: ", round(accuracy_score(y_test, y_test_pred), 3))
    print("Balanced accuracy on test data: ", round(balanced_accuracy_score(y_test, y_test_pred), 3))

# Function to generate classification reports
def print_classification_reports(ytrain_pred, ytrain, ytest_pred, ytest, class_names):
    # Training set
    print("Training Set")
    print(classification_report(ytrain, ytrain_pred, target_names=class_names, digits=3))
    print("------------------")

    # Test set
    print("Test Set")
    print(classification_report(ytest, ytest_pred, target_names=class_names, digits=3))

# Function to change from multi-class to one-vs-rest format
def one_vs_rest(y, id_of_interest):
    res = np.zeros(len(y), dtype=int)
    # Set all classes of interest to 1; the rest remain 0
    res[np.where(y == id_of_interest)[0]] += 1
    return res

# Function to plot the ROC curve
def plot_roc_auc(y_train_pred, y_train, y_test_pred, y_test, class_names, model_name):
    # Consider classes that are common (5), uncommon (6), and somewhat common (3)
    class_ids = [5, 6, 3]

    # Plotting figure and axes
    fig, ax = plt.subplots(len(class_ids), 2, figsize=(12, 18))

    # Create ROC curve for each class on training and test sets
    for i, id in enumerate(class_ids):
        # Training set
        RocCurveDisplay.from_predictions(
            one_vs_rest(y_train, id),
            one_vs_rest(y_train_pred, id),
            name=f"{class_names[id]} vs. rest",
            ax=ax[i][0]
        )

        # Test set
        RocCurveDisplay.from_predictions(
            one_vs_rest(y_test, id),
            one_vs_rest(y_test_pred, id),
            name=f"{class_names[id]} vs. rest",
            ax=ax[i][1]
        )

        # Adding comparison to random classifier
        ax[i][0].plot([0, 1], [0, 1], "r--", label="Random classifier (AUC = 0.5)")
        ax[i][1].plot([0, 1], [0, 1], "r--", label="Random classifier (AUC = 0.5)")

        # Axes labels
        ax[i][0].set_xlabel("False Positive Rate (FPR)")
        ax[i][0].set_ylabel("True Positive Rate (TPR)")
        ax[i][1].set_xlabel("False Positive Rate (FPR)")
        ax[i][1].set_ylabel("True Positive Rate (TPR)")

        # Axes titles
        ax[i][0].set_title(f"Class {class_names[id]}, Frequency: {class_freqs[id]}%")
        ax[i][1].set_title(f"Class {class_names[id]}, Frequency: {class_freqs[id]}%")

        # Axes legends
        ax[i][0].legend()
        ax[i][1].legend()

    # Figure title
    fig.suptitle(f"ROC Curve for {model_name}", fontsize=16)
    plt.show()
```

### B.1.b Writing general code for training CNNs

```python
# CNN hyperparameters
learning_rate = 1e-3
batch_size = 512
epochs = 100
```

```python
# Preparing data for training using CNNs
# Note: since CNNs are usually trained on images, they accept data of the form (batch_size, num_freq, num_time_windows, channels) rather than 1d data
#       Here num_freq = 128
#            num_time_windows = 87
#            channels = 1 (only one value, we are not dealing with rgb images)

# We treat the mel spectograms as grayscale "images"

# Setting the device to run the model on (we are using GPU on colab)
device = torch.device("cuda:0") if torch.cuda.is_available() else torch.device("cpu")

# Reshaping data to be of the form (batch_size, num_freq, num_time_windows, channels)
Xtrain_mel_cnn = Xtrain_mel.reshape(Xtrain_mel.shape[0], 128, 87, 1)
Xtest_mel_cnn = Xtest_mel.reshape(Xtest_mel.shape[0], 128, 87, 1)

# Converting data to torch tensors
Xtrain_mel_cnn_torch = torch.from_numpy(Xtrain_mel_cnn).float().to(device)
ytrain_mel_torch = torch.from_numpy(ytrain_mel)
Xtest_mel_cnn_torch = torch.from_numpy(Xtest_mel_cnn).float().to(device)
ytest_mel_torch = torch.from_numpy(ytest_mel)

#reshaping Xtrain and Xtest to be of the form (batch_size, channels, num_freq, num_time_windows)
Xtrain_mel_cnn_torch = Xtrain_mel_cnn_torch.permute(0, 3, 1, 2)
Xtest_mel_cnn_torch = Xtest_mel_cnn_torch.permute(0, 3, 1, 2)

# One-hot encoding the outputs
ytrain_mel_torch = F.one_hot(ytrain_mel_torch, num_classes=10).float().to(device)
ytest_mel_torch = F.one_hot(ytest_mel_torch, num_classes=10).float().to(device)

#reshaping ytrain and ytest to be of the form (batch_size, num_classes)
ytrain_mel_torch = ytrain_mel_torch.permute(0, 2, 1).squeeze()
ytest_mel_torch = ytest_mel_torch.permute(0, 2, 1).squeeze()

# Creating datasets from the data
train_dataset = torch.utils.data.TensorDataset(Xtrain_mel_cnn_torch, ytrain_mel_torch)
test_dataset = torch.utils.data.TensorDataset(Xtest_mel_cnn_torch, ytest_mel_torch)

# Creating dataloaders from the datasets
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=batch_size, shuffle=True)
```

```python
# Next, we write the code for training a testing a pytorch model. This will be used in both parts B.1.b and B.1.c

# Code for plotting training loss
def plot_training_loss(training_loss, module_name):
    plt.plot(training_loss, label="Training loss")
    plt.title(f"Training loss for {module_name}")
    plt.legend()
    plt.show()

# Code for training a pytorch model
def train(model, train_loader, criterion, optimizer, model_name, epochs=10):
    print("=========================================")
    print("===== Training the model on training set =====")
    model.train()
    training_loss = []
    for epoch in tqdm(range(epochs), total= epochs):
        print("\n")
        running_loss = 0.0
        for i, data in enumerate(train_loader, 0):
            inputs, labels = data
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
            training_loss.append(loss.item())
            if i % 10 == 9:
                print(f'[{epoch + 1}, {i + 1}] loss: {(running_loss / 10):.3f}')
                running_loss = 0.0
    torch.save(model, f"{model_name}.pt")
    plot_training_loss(training_loss, model_name)
    return np.array(training_loss)

# Code for testing a pytorch model
def test(model, test_loader, criterion):
    print("=========================================")
    print("===== Testing the model on test set =====")
    model.eval()
    testing_loss = []
    test_loss = 0
    correct = 0
    with torch.no_grad():
        for data, target in tqdm(test_loader, total=len(test_loader)):
            output = model(data)
            loss = criterion(output, target).item()
            test_loss += loss
            testing_loss.append(loss)
            pred_classes = output.argmax(dim=1, keepdim=True)
            target_classes = target.argmax(dim=1, keepdim=True)
            correct += pred_classes.eq(target_classes).sum().item()
    test_loss /= len(test_loader.dataset)
    print(f'Test set: Average loss: {test_loss:.4f}, Accuracy: {correct}/{len(test_loader.dataset)} ({100. * correct / len(test_loader.dataset):.0f}%)')
    return np.array(testing_loss)

# Code for evaluating the model, displaying accuracies and roc curves
def evaluate(model, train_loader, test_loader, model_name):
    # Creating hard classifications using the CNN model
    model.eval()
    ytrain_pred = []
    ytrain = []
    ytest_pred = []
    ytest = []
    with torch.no_grad():
        print("================================================")
        print("===== Evaluating the model on training set =====")
        for data, target in tqdm(train_loader, total=len(train_loader)):
            output = model(data)
            pred_classes = output.argmax(dim=1, keepdim=True)
            ytrain_pred += pred_classes.cpu().numpy().flatten().tolist()
            actual_classes = target.argmax(dim=1, keepdim=True)
            ytrain += actual_classes.cpu().numpy().flatten().tolist()
        
        print("===============================================")
        print("===== Evaluating the model on test set ========")
        for data, target in tqdm(test_loader, total=len(test_loader)):
            output = model(data)
            pred_classes = output.argmax(dim=1, keepdim=True)
            ytest_pred += pred_classes.cpu().numpy().flatten().tolist()
            actual_classes = target.argmax(dim=1, keepdim=True)
            ytest += actual_classes.cpu().numpy().flatten().tolist()
        
    # Displaying accuracies
    print("===================================================")
    print("============== Displaying Accuracies ==============")
    print_accuracies(ytrain_pred, ytrain, ytest_pred, ytest)

    # Displaying Classification Report
    print("===================================================")
    print("======= Displaying Classification Report ==========")
    print_classification_reports(ytrain_pred, ytrain, ytest_pred, ytest, class_names)
    
    # Displaying ROC curves
    plot_roc_auc(ytrain_pred, ytrain, ytest_pred, ytest, class_names, model_name)
```

### B.1.c Simple CNN model from scratch (using PyTorch)

#### Model Architecture

```python
# We create a simple CNN model from scratch using pytorch (we use mel spectogram data as input)
# We used this pytorch example on CNNs for reference: https://github.com/pytorch/examples/blob/main/mnist/main.py

class CNN_basic(nn.Module):
    def __init__(self):
        super(CNN_basic, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, 3, padding=1)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(32, 64, 3, padding=1)
        self.fc1 = nn.Linear(64 * 32 * 21, 128)
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 64 * 32 * 21)
        x = F.relu(self.fc1(x))
        x = F.softmax(self.fc2(x), dim=1)
        return x

CNN_basic_model = CNN_basic().to(device)
summary(CNN_basic_model, input_size=(1, 128, 87))
```

#### Training and Testing

```python
# Setting training settings (loss function (i.e criterion) and optimization algorithm)
# For optimization we use Adam, which is a widely-used optimization algorithm
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(CNN_basic_model.parameters(), lr=learning_rate)

# Training the CNN_basic_model
CNN_basic_training_loss = train(CNN_basic_model, train_loader, criterion, optimizer, "CNN_basic_model", epochs=epochs)
# Testing the CNN_basic_model
CNN_basic_testing_loss = test(CNN_basic_model, test_loader, criterion)
```

#### Displaying Accuracies, Classification Reports and ROC Curves

```python
evaluate(CNN_basic_model, train_loader, test_loader, "CNN_basic_model")
```

### B.2. More Complicated Models—Hyperparameter Tuning and Validation

#### Using Grid Search and K-Fold Cross Validation for Hyperparameter Selection

```python
# SciPy randint defines a discrete uniform distribution
n_estimators_dist = randint(80, 140)
max_depth_dist = randint(30, 70)

# Dictionary with parameter distributions for randomized search
param_dist = {
    "n_estimators": n_estimators_dist,
    "max_depth": max_depth_dist
}

# Initialize a random forest classifier class
rfc = RandomForestClassifier(criterion="log_loss")

# Number of folds or splits of the data for k-fold cross validation
k_folds = 3

# Perform randomized search CV with 5-fold cross-validation
rand_search = RandomizedSearchCV(
    rfc,
    param_distributions=param_dist,
    cv=k_folds,
    n_iter=10,
    scoring="accuracy",
    verbose=3
)

# Fit the randomized search CV model and print best parameter values found
rand_search.fit(Xtrain_mel, ytrain_mel.reshape(ytrain_mel.shape[0],))
print("Best Hyperparameters:", rand_search.best_params_)
```

```python
# Helper function to generate a set of values centered around a given value
def gen_range(val, num, step, min_val=1):
  # This function generates a range concentrated around the value
  dist_floor = floor(num / 2.) * step
  dist_ceil = ceil(num/ 2.) * step
  return range(max(val - dist_floor, min_val), val + dist_ceil, step)

# Initialize a random forest classifier class instance
rfc = RandomForestClassifier()

# Generate a hyperparameter spece (grid) of values centered around provided values
param_grid = {
    "n_estimators": gen_range(rand_search.best_params_["n_estimators"], 2, 2),
    "max_depth": gen_range(rand_search.best_params_["max_depth"], 2, 1),
  }

# Perform GridSearch on the defined grid with k-fold cross-validation
k_folds = 5
gs = GridSearchCV(
    rfc,
    param_grid=param_grid,
    cv=k_folds,
    scoring="accuracy",
    verbose=3
)

# Fit the GridSearch model and print best parameter values found
gs.fit(Xtrain_mel, ytrain_mel.reshape(ytrain_mel.shape[0],))
print("Best Hyperparameters:", gs.best_params_)
```

```python
print(f"Best hyperparameters: {gs.best_params_}")
print(f"Score: {gs.best_score_}")
np.save("gs_best_params.npy", gs.best_params_)
np.save("gs_best_score.npy", gs.best_score_)
np.save("gs_best_cv.npy", gs.cv_results_)
```

## Part C. CNN Pretrained on ImageNet (Resnet Model)

#### C.1 Data Augmentation

```python
# Data Augmentation
# We noticed that the pre-trained CNN performs worse than the simpler CNN, despite what the paper (https://arxiv.org/pdf/2007.11154.pdf) claims.
# So, we decided to augment our data (i.e increase data size) since we assumed that fine-tuning a larger CNN, whose weights are already pre-trained
# on a different unrelated dataset would require us to use more data (to make sure the pre-trained model is appropriately fine-tuned for the new task).
# Note that the paper does this but in a different way

# frequency masking
def freq_mask(X, F=16):
  # Inputs:
  # X: batch_size x 1 x 178 x 87 mel spec data
  # F: the maximum width of each of the frequency masks
  #
  # Returns: batch_size x 1 x 178 x 87 mel spec data with frequency masks applied
  
  for i in range(X.shape[0]):
    spec = X[i, :, :, :]
    freq_mask = FrequencyMasking(freq_mask_param = F)
    spec_freq_masked = freq_mask(spec)
    X[i, :, :, :] = spec_freq_masked
  return X

# time masking
def time_mask(X, T=8):
  # Inputs:
  # X: batch_size x 1 x 178 x 87 mel spec data
  # T: the maximum width of each of the time masks
  #
  # Returns: batch_size x 1 x 178 x 87 mel spec data with time masks applied
  
  for i in range(X.shape[0]):
    spec = X[i, :, :, :]
    time_mask = TimeMasking(time_mask_param = T)
    spec_time_masked = time_mask(spec)
    X[i, :, :, :] = spec_time_masked
  return X

# frequency and time masking
def freq_time_mask(spec, F=30, T=40):
  spec = freq_mask(spec, F=F) if F else spec
  spec = time_mask(spec, T=T) if T else spec
  return spec

def mel_augment(X, Y, freq_mask_window=16, time_mask_window=8, num_masked_copies=1):
  # This function performs mel augmentation using the method described (masking frequency and time windows and adding
  # the newly-generated data to the training set)
  # here: https://towardsdatascience.com/audio-deep-learning-made-simple-sound-classification-step-by-step-cebc936bbe5
  
  print(f"Augmenting the mel spectogram data by {num_masked_copies} masks per sample")
  list_augmented_copies = [X]
  for _ in tqdm(range(num_masked_copies), total=num_masked_copies):
    # creating a clone
    X_clone = X.clone()

    # applying frequency masking
    X_freq_masked = freq_mask(X_clone, F=freq_mask_window) if freq_mask_window else X_clone

    # applying time masking
    X_time_masked = time_mask(X_clone, T=time_mask_window) if time_mask_window else X_clone

    # applying both frequency and time masking
    X_freq_time_masked = freq_time_mask(X_clone, F=freq_mask_window, T=time_mask_window) if freq_mask_window and time_mask_window else X_clone

    list_augmented_copies += [X_freq_masked, X_time_masked, X_freq_time_masked]

  # augmenting the features
  X = torch.cat(tuple(list_augmented_copies), dim=0)

  # augmenting the labels
  Y = torch.cat((Y,) * len(list_augmented_copies), dim=0)

  return X, Y
```

```python
print(f"Number of datapoints before augmenting={Xtrain_mel_cnn_torch.shape[0]}")

Xtrain_mel_cnn_torch_aug, ytrain_mel_torch_aug = mel_augment(Xtrain_mel_cnn_torch, ytrain_mel_torch, num_masked_copies=3)

print(f"\nNumber of datapoints after augmenting={Xtrain_mel_cnn_torch_aug.shape[0]}")
print(f"The number of datapoints was augmented by {Xtrain_mel_cnn_torch_aug.shape[0] / Xtrain_mel_cnn_torch.shape[0]}")
```

```python
# Creating datasets from the augmented data
train_dataset_aug = torch.utils.data.TensorDataset(Xtrain_mel_cnn_torch_aug, ytrain_mel_torch_aug)

# Creating dataloaders from the augmented datasets
train_loader_aug = torch.utils.data.DataLoader(train_dataset_aug, batch_size=batch_size, shuffle=True)
```

#### C.2 Pulling Resnet Model

```python
# Here we use a technique called transfer-learning to load a pre-trained model (resnet18)
# and make the appropriate changes to make sure the model fits to our use-case:
# i.e we train the last layer to classify the 10 classes

# Inspiration for code: https://pytorch.org/tutorials/beginner/finetuning_torchvision_models_tutorial.html
# Idea of using pretrained CNNs for audio classification: https://arxiv.org/pdf/2007.11154.pdf

# Loading the pretrained resnet model
CNN_resnet_model = models.resnet18(weights=models.ResNet18_Weights.DEFAULT).cuda()
```

```python
# Showing the resnet layers to decide on what layers to edit

# Notice that vision models like resnet usually accept 3 channels for input (rgb)
# instead of 1 channel which we have for our mel spectogram data
summary(CNN_resnet_model, input_size=(3, 128, 87))
```

Notice that the first layer accepts in 3 channels, wheras we only need 1, so we need to change that. The last linear layer has 1000 classes, but we want only 10, so we adjust that too. We leave the weights for the remaining layers unchanged because we do not want to overfit the model (the resnet architecture is too large for our limited data)

#### C.3 Editing Resnet layers and fine-tuning

```python
# Editing the pre-trained model layers

# Freezing all the layers except the last layer
for param in CNN_resnet_model.parameters():
    param.requires_grad = False

# Changing the first layer to accept just 1 channel
CNN_resnet_model.conv1 = nn.Conv2d(1, 64, kernel_size=(7, 7), stride=(2, 2), padding=(3, 3), bias=False)

# Changing the last layer to classify the 10 classes and a softmax activation (to get probabilities)
CNN_resnet_model.fc = nn.Sequential(
    nn.Linear(512, 10),
    nn.Softmax(dim=1)
).cuda()

# Running model on gpu
_ = CNN_resnet_model.to(device)
```

```python
# Setting training settings (loss function (i.e criterion) and optimization algorithm)
# For optimization we use Adam, which is a widely-used optimization algorithm
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(CNN_resnet_model.parameters(), lr=learning_rate)

# Fine-tuning the resent model on the augmented mel spectogram data
CNN_resnet_training_loss = train(CNN_resnet_model, train_loader_aug, criterion, optimizer, "CNN_resnet_model", epochs=epochs)

# Testing the model
CNN_resnet_testing_loss = test(CNN_resnet_model, test_loader, criterion)
```

#### Displaying Accuracies, Classification Reports and ROC Curves
